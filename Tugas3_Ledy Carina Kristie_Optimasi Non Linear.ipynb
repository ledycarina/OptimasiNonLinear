# Soal 1 (Konsep Dasar)

Diberikan $f(x)=x^2$.

## a) Hitung gradien (turunan pertama)

1. Gunakan aturan turunan untuk pangkat: $\dfrac{d}{dx} x^n = n x^{n-1}$.
2. Untuk $f(x)=x^2$:

$$
f'(x)=\frac{d}{dx}x^2 = 2x.
$$

**Jadi gradien (turunan pertama) adalah $f'(x)=2x$.**



## b) Rumus iterasi Gradient Descent dengan learning rate $\alpha$

Gradient descent umum untuk satu variabel:

$$
x_{k+1} = x_k - \alpha \, f'(x_k).
$$

Karena $f'(x)=2x$, masukkan ke rumus:

$$
x_{k+1} = x_k - \alpha\,(2x_k) = x_k(1 - 2\alpha).
$$

Sebagai catatan: iterasi umum setelah $n$ langkah:

$$
x_n = (1-2\alpha)^n \, x_0.
$$

Untuk konvergensi ke 0 diperlukan $|1-2\alpha|<1$ âŸº $0<\alpha<1$ (untuk fungsi kuadrat sederhana ini).



## c) Tentukan $x_1$ jika $x_0=4$ dan $\alpha=0.1$

Cara 1 â€” langsung dari rumus update:

$$
x_{1} = x_0 - \alpha f'(x_0).
$$

Langkah perhitungan digit-by-digit:

1. Hitung gradien di $x_0$: $f'(4)=2\cdot 4=8$.
2. Kalikan dengan $\alpha$: $\alpha\cdot f'(4)=0.1\cdot 8$.
   â€” $0.1 = \dfrac{1}{10}$, jadi $0.1\cdot 8 = \dfrac{8}{10} = 0.8$.
3. Update: $x_1 = 4 - 0.8 = 3.2$.

Cara 2 â€” pakai bentuk faktor $(1-2\alpha)x_0$:

1. Hitung faktor: $1 - 2\alpha = 1 - 2(0.1) = 1 - 0.2 = 0.8$.
2. Kalikan: $x_1 = 0.8 \cdot 4$.
   â€” $0.8 = \dfrac{8}{10}$, jadi $\dfrac{8}{10}\cdot 4 = \dfrac{32}{10} = 3.2$.

**Jadi $x_1 = 3.2$.**

(Optional â€” cek bahwa fungsi menurun)

* $f(x_0)=f(4)=4^2=16$.
* $f(x_1)=f(3.2)=3.2^2$. Hitung: $3.2 = \dfrac{32}{10}$ â†’ $3.2^2=\dfrac{32^2}{100}=\dfrac{1024}{100}=10.24$.
  Karena $10.24 < 16$, langkah gradient descent berhasil menurunkan nilai fungsi.



## **Soal 1 (Konsep Dasar)**

Diberikan fungsi objektif:

$$
f(x) = x^2
$$

### **a. Hitung gradien (turunan pertama) dari fungsi tersebut**

1. **Definisi turunan**:
   Jika $f(x) = x^n$, maka:

   $$
   \frac{d}{dx} x^n = n \cdot x^{n-1}.
   $$

2. Karena fungsi kita adalah $f(x) = x^2$, maka:

   $$
   f'(x) = \frac{d}{dx}(x^2) = 2x.
   $$

ðŸ‘‰ **Keterangan**:

* $f'(x)$ disebut **gradien** atau **turunan pertama**.
* Gradien ini menunjukkan arah dan besarnya perubahan fungsi $f(x)$ terhadap $x$.



### **b. Rumus iterasi Gradient Descent dengan learning rate $\alpha$**

1. **Prinsip gradient descent**:
   Untuk meminimalkan fungsi, kita melakukan update nilai $x$ dengan mengurangi gradien dikali *learning rate*.

   Rumus umum:

   $$
   x_{k+1} = x_k - \alpha f'(x_k).
   $$

2. Karena $f'(x) = 2x$, maka:

   $$
   x_{k+1} = x_k - \alpha (2x_k).
   $$

3. Sederhanakan:

   $$
   x_{k+1} = (1 - 2\alpha)x_k.
   $$

ðŸ‘‰ **Keterangan**:

* $\alpha$ = *learning rate* (ukuran langkah).
* $(1-2\alpha)$ adalah faktor pengurang yang menentukan seberapa cepat nilai $x$ mendekati minimum (yaitu $x=0$).
* Agar metode konvergen, syaratnya $0 < \alpha < 1$.



### **c. Tentukan $x_1$ jika $x_0=4$ dan $\alpha=0.1$**

1. Gunakan rumus update:

   $$
   x_{1} = x_0 - \alpha f'(x_0).
   $$

2. Hitung gradien di titik $x_0=4$:

   $$
   f'(4) = 2(4) = 8.
   $$

3. Kalikan dengan $\alpha = 0.1$:

   $$
   \alpha f'(4) = 0.1 \times 8 = 0.8.
   $$

4. Update nilai:

   $$
   x_1 = 4 - 0.8 = 3.2.
   $$

ðŸ‘‰ **Keterangan**:

* $x_1 = 3.2$ adalah nilai baru setelah satu langkah *gradient descent*.
* Nilai fungsi juga berkurang:

  $$
  f(x_0) = f(4) = 16, \quad f(x_1) = f(3.2) = 10.24.
  $$

  Artinya, iterasi berhasil menurunkan nilai fungsi.



âœ… **Ringkasan Akhir**

* Turunan pertama: $f'(x) = 2x$.
* Rumus update gradient descent: $x_{k+1} = (1-2\alpha)x_k$.
* Dengan $x_0=4, \alpha=0.1$, diperoleh $x_1 = 3.2$.





### ðŸ”‘ **Istilah-Asing & Penjelasannya**

1. **Fungsi Objektif (Objective Function)**

   * Adalah fungsi matematika yang ingin kita *minimalkan* atau *maksimalkan*.
   * Dalam soal: $f(x)=x^2$.
   * Disebut *objective* karena menjadi tujuan utama optimisasi.

---

2. **Gradien (Gradient)**

   * Gradien adalah turunan pertama dari fungsi.
   * Menunjukkan arah perubahan fungsi:

     * Jika gradien positif â†’ fungsi naik.
     * Jika gradien negatif â†’ fungsi turun.
   * Dalam soal: gradiennya adalah $f'(x)=2x$.

---

3. **Turunan Pertama (First Derivative)**

   * Operasi matematika untuk menghitung laju perubahan fungsi terhadap variabel.
   * Contoh: turunan dari $x^2$ adalah $2x$.
   * Digunakan dalam optimisasi untuk menemukan titik minimum/maksimum.

---

4. **Gradient Descent**

   * Metode optimisasi (algoritma) untuk mencari nilai minimum dari suatu fungsi.
   * Caranya: bergerak ke arah yang berlawanan dengan gradien, sedikit demi sedikit.
   * Disebut â€œ*descent*â€ (turun) karena kita bergerak menuruni kurva fungsi sampai ke titik terendah.

---

5. **Iterasi (Iteration)**

   * Proses pengulangan langkah update sampai nilai fungsi mendekati optimum.
   * Contoh: dari $x_0=4$ â†’ dihitung $x_1=3.2$ â†’ lalu $x_2, x_3,$ dst.

---

6. **Learning Rate ($\alpha$)**

   * Parameter yang menentukan seberapa besar langkah yang diambil dalam setiap iterasi.
   * Jika $\alpha$ terlalu besar â†’ bisa lompat-lompat dan tidak konvergen.
   * Jika $\alpha$ terlalu kecil â†’ prosesnya sangat lambat.
   * Dalam soal: $\alpha=0.1$.

---

7. **Konvergen (Convergence)**

   * Kondisi ketika nilai iterasi semakin dekat ke titik optimal (minimum).
   * Untuk fungsi $f(x)=x^2$, titik minimumnya ada di $x=0$.
   * Proses iterasi seiring waktu akan mendekati $x=0$.




```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi objektif
def f(x):
    return x**2

# Gradien (turunan pertama)
def grad_f(x):
    return 2*x

# Parameter awal
x0 = 4           # titik awal
alpha = 0.1      # learning rate
n_iter = 5       # jumlah iterasi untuk ilustrasi

# Simpan hasil iterasi
x_vals = [x0]
f_vals = [f(x0)]

x = x0
for i in range(n_iter):
    # update dengan rumus gradient descent
    x = x - alpha * grad_f(x)
    x_vals.append(x)
    f_vals.append(f(x))

# Plot grafik fungsi
x_plot = np.linspace(-5, 5, 200)
y_plot = f(x_plot)

plt.figure(figsize=(8,6))
plt.plot(x_plot, y_plot, label="f(x) = x^2")   # kurva fungsi
plt.scatter(x_vals, f_vals, color="red", zorder=5)  # titik iterasi
plt.plot(x_vals, f_vals, color="red", linestyle="--", label="Langkah Gradient Descent")

# Tambahkan label pada titik
for i, (xv, yv) in enumerate(zip(x_vals, f_vals)):
    plt.text(xv, yv+0.5, f"x{i}={xv:.2f}", fontsize=9)

plt.title("Ilustrasi Gradient Descent pada f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
