visualisasi gambar dari f(x,y) =sin(x) + cos(y)

````python
import numpy as np
import matplotlib.pyplot as plt

# Create a meshgrid for x and y values
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.linspace(-2 * np.pi, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)

# Calculate the function values
Z = np.sin(X) + np.cos(Y)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
ax.set_title('Plot of f(x,y) = sin(x) + cos(y)')

# Add a color bar
fig.colorbar(surf)

# Show the plot
plt.show()
````

# Secara matematis
Fungsi yang dipetakan adalah:

$$
f(x,y) = \sin(x) + \cos(y)
$$

* **Sumbu X dan Y**: variabel bebas $x$ dan $y$.
* **Sumbu vertikal (Z / f(x,y))**: nilai fungsi $\sin(x) + \cos(y)$.
* **Warna (color bar di kanan)**: menunjukkan besar nilai fungsi.

  * Warna **kuning**  nilai mendekati maksimum (+2).
  * Warna **ungu tua**  nilai mendekati minimum (2).


##  Cara membaca bentuk permukaan

1. Fungsi ini **bergelombang** karena ada sinus dan cosinus.
2. Gelombang terlihat **berulang (periodik)** ke arah sumbu X maupun Y.

   * Sepanjang sumbu X, pola berulang setiap $2\pi$.
   * Sepanjang sumbu Y, pola juga berulang setiap $2\pi$.
3. Permukaan terdiri dari **puncak (maksimum lokal)** dan **lembah (minimum lokal)** yang tersusun rapi seperti ubin.



##  Titik penting

* **Nilai maksimum** terjadi saat $\sin(x)=1$ dan $\cos(y)=1$.

  * Contoh: $x=\frac{\pi}{2}, y=2k\pi$  $f=2$.
* **Nilai minimum** terjadi saat $\sin(x)=-1$ dan $\cos(y)=-1$.

  * Contoh: $x=\frac{3\pi}{2}, y=\pi+2k\pi$  $f=-2$.
* Di antara puncak dan lembah, ada lembah dangkal dan bukit kecil  ini membuat permukaan terlihat seperti gelombang laut dari atas.


##  Kesimpulan

* Grafik menunjukkan bahwa fungsi **periodik**, selalu berulang.
* **Puncak (warna kuning)** = nilai besar, **lembah (warna ungu tua)** = nilai kecil.
* Fungsi tidak punya satu maksimum/minimum global saja, tapi **banyak** karena sinus dan cosinus terus berulang.



 Jadi, : bayangkan grafik ini seperti **permukaan laut bergelombang** dengan bukit (maksimum) dan lembah (minimum) yang berulang sepanjang sumbu X dan Y.

# Visualisasi untuk gambar a

````python
import numpy as np
import matplotlib.pyplot as plt

# Definisi fungsi a)
def f_a(x1, x2, x3):
    return 2*x1**2 + 3*x2**2 + 3*x1*x3

# Buat grid untuk x1 dan x2, dengan x3 = 0 (slice agar mudah divisualisasikan)
x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
X3 = 0  # kita ambil irisan di x3=0
Z = f_a(X1, X2, X3)

# Plot
fig = plt.figure(figsize=(12,5))

# Surface plot (3D)
ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.plot_surface(X1, X2, Z, cmap='viridis', edgecolor='none')
ax1.set_title("Permukaan f(x1,x2,x3) dengan x3=0")
ax1.set_xlabel("x1")
ax1.set_ylabel("x2")
ax1.set_zlabel("f")

# Contour plot (2D)
ax2 = fig.add_subplot(1,2,2)
contour = ax2.contourf(X1, X2, Z, levels=30, cmap='viridis')
plt.colorbar(contour, ax=ax2)
ax2.set_title("Contour plot f(x1,x2,x3) (x3=0)")
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")

plt.tight_layout()
plt.show()
````


##  **Analisis Matematis**

1. **Turunan parsial (gradien):**

   $$
   \frac{\partial f}{\partial x_1} = 4x_1 + 3x_3,\quad
   \frac{\partial f}{\partial x_2} = 6x_2,\quad
   \frac{\partial f}{\partial x_3} = 3x_1
   $$

2. **Titik stasioner:**
   Set semua turunan = 0:

   $$
   4x_1+3x_3=0,\quad 6x_2=0,\quad 3x_1=0
   $$

    dari $\frac{\partial f}{\partial x_2}=0 \Rightarrow x_2=0$
    dari $\frac{\partial f}{\partial x_3}=0 \Rightarrow x_1=0$
    substitusi ke $\frac{\partial f}{\partial x_1}$: $0+3x_3=0 \Rightarrow x_3=0$.
   Jadi titik stasioner = $(0,0,0)$.

3. **Hessian matriks:**

   $$
   H=\begin{bmatrix}
   \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1\partial x_2} & \frac{\partial^2 f}{\partial x_1\partial x_3} \\
   \frac{\partial^2 f}{\partial x_2\partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \frac{\partial^2 f}{\partial x_2\partial x_3} \\
   \frac{\partial^2 f}{\partial x_3\partial x_1} & \frac{\partial^2 f}{\partial x_3\partial x_2} & \frac{\partial^2 f}{\partial x_3^2}
   \end{bmatrix}
   =
   \begin{bmatrix}
   4 & 0 & 3 \\
   0 & 6 & 0 \\
   3 & 0 & 0
   \end{bmatrix}
   $$

    Nilai eigen dari Hessian: ada positif dan ada negatif  Hessian **indefinite**.
    Artinya titik (0,0,0) adalah **saddle point**, **bukan minimum atau maksimum lokal**.

4. **Kesimpulan:**

   * Fungsi ini **tidak punya maksimum terbatas** (jika $x_1,x_2$ besar  $f \to \infty$).
   * Minimum pun tidak ada (karena $x_1x_3$ bisa memberi nilai negatif sangat besar).
   * Jadi fungsi ini **tak terbatas (unbounded)**.

f(x1,x2,x3)=2x1+3x2+3x3

````python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi tujuan b)
def f_b(x1, x2, x3):
    return 2*x1 + 3*x2 + 3*x3

# Buat grid untuk x1 dan x3
x1 = np.linspace(0, 3, 100)
x3 = np.linspace(0, 5, 100)
X1, X3 = np.meshgrid(x1, x3)
X2 = (3 - X1)/2  # dari kendala x1+2x2=3

# Hitung nilai fungsi
Z = f_b(X1, X2, X3)

# Masking daerah feasible: X2>=0, X1+X3>=3
mask = (X2 >= 0) & (X1 + X3 >= 3)
Z_masked = np.where(mask, Z, np.nan)

# Plot contour
plt.figure(figsize=(7,6))
cp = plt.contourf(X1, X3, Z_masked, levels=30, cmap="plasma")
plt.colorbar(cp, label="f(x1,x2,x3)")
plt.xlabel("x1")
plt.ylabel("x3")
plt.title("Soal b) Contour plot f(x1,x2,x3) pada daerah feasible")

# Gambarkan batas kendala
plt.plot([0,3],[3,0], 'w--', label="x1+x3=3 (kendala aktif)")
plt.legend()
plt.show()
````


#  Soal b

Fungsi tujuan:

$$
f(x_1,x_2,x_3) = 2x_1 + 3x_2 + 3x_3
$$

Kendala:

$$
\begin{aligned}
x_1 + 2x_2 &= 3 \quad \text{(kendala persamaan)} \\
x_1 + x_3 &\ge 3 \quad \text{(kendala pertidaksamaan)} \\
x_1, x_2, x_3 &\ge 0 \quad \text{(non-negativitas)}
\end{aligned}
$$



#  Penyelesaian Matematis

1. **Gunakan kendala persamaan:**

   $$
   x_2 = \frac{3 - x_1}{2}, \quad \text{dengan syarat } x_1 \le 3.
   $$

2. **Substitusi ke fungsi tujuan:**

   $$
   f(x_1,x_3) = 2x_1 + 3\left(\frac{3-x_1}{2}\right) + 3x_3
   $$

   $$
   f(x_1,x_3) = 2x_1 + \frac{9}{2} - \frac{3}{2}x_1 + 3x_3
   $$

   $$
   f(x_1,x_3) = \frac{1}{2}x_1 + \frac{9}{2} + 3x_3
   $$

3. **Kendala tersisa:**

   * $x_1+x_3 \ge 3$
   * $0 \le x_1 \le 3$
   * $x_2=(3-x_1)/2 \ge 0 \Rightarrow x_1 \le 3$.

4. **Interpretasi:**
   Fungsi tujuan linear  nilai minimum dicapai di titik batas feasible region.

   * Jika $x_1=0$, maka $x_2=1.5$, kendala $x_1+x_3 \ge 3 \Rightarrow x_3 \ge 3$.
      $f=0.5(0)+4.5+3(3)=13.5$.
   * Jika $x_1=3$, maka $x_2=0$, kendala $x_1+x_3 \ge 3 \Rightarrow x_3 \ge 0$.
     Ambil $x_3=0$.
      $f=0.5(3)+4.5+0=6.0$.  ini lebih kecil.

 Jadi **minimum global** ada di:

$$
(x_1,x_2,x_3) = (3,0,0), \quad f_{\min}=6
$$



#  Interpretasi Grafik

* Area berwarna = **daerah feasible** (memenuhi kendala).
* Garis kontur = nilai fungsi $f$.
* **Titik merah** = titik optimum $(3,0,0)$.
* Nilai minimum $f=6$.

### a) Kurangi variabel dengan substitusi

Dari kendala $x_1+x_2=4$  kita bisa tulis:

$$
x_2 = 4 - x_1
$$

Maka fungsi tujuan jadi:

$$
f(x_1,x_3) = 2x_1 + 3(4-x_1)x_3
$$

$$
= 2x_1 + 12x_3 - 3x_1x_3
$$

### b) Kendala yang tersisa

* $x_1x_3 \geq 3$
* $x_1 \geq 0, \; x_2=4-x_1 \geq 0 \Rightarrow 0 \leq x_1 \leq 4$
* $x_3 \geq 0$

### c) Analisis sifat fungsi

Fungsi $f(x_1,x_3) = 2x_1 + 12x_3 - 3x_1x_3$ adalah **non-linear** (ada perkalian $x_1x_3$).

### d) Substitusi batas $x_1x_3=3$

Dari $x_3 = \frac{3}{x_1}$, masukkan ke fungsi:

$$
f(x_1) = 2x_1 + 12\left(\frac{3}{x_1}\right) - 3x_1\left(\frac{3}{x_1}\right)
$$

$$
= 2x_1 + \frac{36}{x_1} - 9
$$

### e) Optimasi satu variabel

Turunan:

$$
f'(x_1) = 2 - \frac{36}{x_1^2}
$$

Set $f'(x_1)=0$:

$$
2 = \frac{36}{x_1^2} \quad \Rightarrow \quad x_1^2 = 18 \quad \Rightarrow \quad x_1 = \sqrt{18} \approx 4.24
$$

Tapi ingat kendala: $0 \leq x_1 \leq 4$. Jadi solusi $4.24$ **tidak feasible**.

 Maka minimum terjadi di **batas** $x_1=4$.

* Jika $x_1=4 \Rightarrow x_2=0$.

#   soal c dengan pendekatan KKT (KarushKuhnTucker)

Fungsi tujuan:

$$
f(x_1,x_2,x_3) = 2x_1 + 3x_2x_3
$$

Kendala:

$$
x_1 + x_2 = 4, \quad x_1x_3 \geq 3, \quad x_1,x_2,x_3 \geq 0
$$



# 1 Bentuk masalah standard

* Fungsi: $f(x)=2x_1+3x_2x_3$ (ingin **minimalkan**)
* Kendala equality:

  $$
  h(x) = x_1+x_2-4=0
  $$
* Kendala inequality:

  $$
  g_1(x) = 3 - x_1x_3 \leq 0, \quad
  g_2(x) = -x_1 \leq 0, \quad
  g_3(x) = -x_2 \leq 0, \quad
  g_4(x) = -x_3 \leq 0
  $$



# 2 Lagrangian

$$
\mathcal{L}(x,\lambda,\mu) = 2x_1 + 3x_2x_3
+ \lambda(x_1+x_2-4)
+ \mu_1(3-x_1x_3) - \mu_2 x_1 - \mu_3 x_2 - \mu_4 x_3
$$

dengan:

* $\lambda$ = multiplier untuk equality,
* $\mu_i \geq 0$ = multiplier untuk inequality.



# 3 Kondisi KKT

### (a) Stationarity

Turunan parsial terhadap variabel = 0:

1. $\frac{\partial L}{\partial x_1} = 2 + \lambda - \mu_2 - \mu_1x_3 = 0$
2. $\frac{\partial L}{\partial x_2} = 3x_3 + \lambda - \mu_3 = 0$
3. $\frac{\partial L}{\partial x_3} = 3x_2 - \mu_4 - \mu_1x_1 = 0$



### (b) Primal feasibility

* $x_1+x_2=4$
* $x_1x_3 \geq 3$
* $x_1,x_2,x_3 \geq 0$



### (c) Dual feasibility

$\mu_1,\mu_2,\mu_3,\mu_4 \geq 0$



### (d) Complementary slackness

* $\mu_1(3-x_1x_3)=0$
* $\mu_2 x_1=0$
* $\mu_3 x_2=0$
* $\mu_4 x_3=0$



# 4 Analisis solusi

* Dari solusi analitik sebelumnya, kandidat optimum:
  $(x_1,x_2,x_3)=(4,0,0.75)$.

Cek kondisi:

* **Primal feasibility:**
  $x_1+x_2=4$ , $4\cdot 0.75=3 \geq 3$ , nonnegatif .

* **Complementary slackness:**

  * Kendala $x_1x_3 \geq 3$ aktif ($=3$)  $\mu_1 \geq 0$.
  * Kendala $x_2 \geq 0$ aktif (x2=0)  $\mu_3 \geq 0$.
  * Kendala lain (x1=4>0, x3=0.75>0) tidak aktif  $\mu_2=\mu_4=0$.

* **Stationarity:**

  1. $2+\lambda -\mu_1(0.75)=0$  $\lambda = -2+0.75\mu_1$
  2. $3(0.75)+\lambda-\mu_3=0$  $\lambda= \mu_3 -2.25$
  3. $3(0) -\mu_1(4)=0$  $\mu_1=0$

Substitusi $\mu_1=0$:

* Dari (1): $\lambda=-2$
* Dari (2): $-2= \mu_3-2.25 \Rightarrow \mu_3=0.25 \geq 0$ 

Semua kondisi KKT terpenuhi. 



# 5 Kesimpulan

* **Solusi optimum:**

  $$
  (x_1,x_2,x_3)=(4,0,0.75)
  $$
* **Nilai minimum:**

  $$
  f(4,0,0.75)=2(4)+3(0)(0.75)=8
  $$
* **Interpretasi:** Fungsi mencapai minimum di **batas kendala** $x_1x_3=3$, saat $x_2=0$.
* **KKT multipliers:** $\mu_1=0,\;\mu_2=0,\;\mu_3=0.25,\;\mu_4=0,\;\lambda=-2$.




# KOSAKATA

* **Turunan / partial**  laju perubahan pada satu arah.
* **Gradient**  vektor semua turunan parsial; arah naik tercepat.
* **Hessian**  matriks turunan kedua; memberi informasi kelengkungan.
* **Contour / Permukaan**  cara visual untuk melihat level-set (kontur) dan bentuk 3D (surface).
* **Pengali Lagrange / KKT**  kondisi untuk menyelesaikan masalah dengan kendala (equality & inequality).
* **Gradient Descent / Newton**  algoritme numerik iteratif untuk menemukan optimum.



# 1) Turunan (1D) dan turunan parsial (multivariat)

**Definisi singkat**

* Turunan 1D: $f'(x)=\lim_{h\to0}\frac{f(x+h)-f(x)}{h}$. Artinya kemiringan garis singgung di titik $x$.
* Turunan parsial untuk $f(x,y)$: $\frac{\partial f}{\partial x}$ adalah laju perubahan $f$ ketika $x$ berubah kecil dan $y$ dianggap tetap.

**Intuisi mudah**

* Turunan memberi tahu apakah fungsi naik atau turun pada arah tertentu. Jika turunan = 0  titik stasioner (kemungkinan min/max/saddle).

**Contoh**

* $f(x,y)=x^2+y^2$. $\partial f/\partial x = 2x,\; \partial f/\partial y = 2y$.



# 2) Gradient

**Definisi**

$$
\nabla f(\mathbf{x}) = \begin{bmatrix}\frac{\partial f}{\partial x_1}\\ \vdots \\ \frac{\partial f}{\partial x_n}\end{bmatrix}.
$$

**Intuisi**

* Gradient menunjuk arah **naik paling cepat** dari fungsi; besarnya ($\|\nabla f\|$) menunjukkan seberapa curam arah itu.
* Pada plot kontur (garis level), gradient selalu **tegak lurus** pada garis kontur.

**Contoh singkat**

* $f(x,y)=x^2+y^2$  $\nabla f = [2x,\;2y]^T$. Di titik (1,1), gradient = (2,2)  arah naik tercepat.



# 3) Hessian (matriks)

**Definisi**

* Hessian adalah matriks turunan kedua:

$$
H_f(\mathbf{x}) = \begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \cdots & \frac{\partial^2 f}{\partial x_1\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n\partial x_1} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}.
$$

**Intuisi**

* Memberi tahu **kecuraman / kelengkungan** fungsi di sekitar titik.
* Jika Hessian **positif definite** di suatu titik stasioner  titik itu **minimum lokal**. Jika **negatif definite**  maksimum lokal. Jika ada campuran tanda eigen  saddle point.

**Cek cepat (22)**

* Untuk matriks simetris $H=\begin{pmatrix}a & b\\ b & c\end{pmatrix}$, positif definite bila $a>0$ dan $\det(H)=ac-b^2>0$.

**Contoh**

* $f=x^2+y^2$  $H = \begin{bmatrix}2&0\\0&2\end{bmatrix}$ (positif definite)  global minimum.



# 4) Contour dan Permukaan (surface)

**Contour (garis level)**

* Garis yang menghubungkan titik-titik $(x,y)$ dengan nilai fungsi sama $f(x,y)=k$. Berguna untuk melihat lokasi minimum/maksimum dan bentuk lembah/puncak pada proyeksi 2D.

**Permukaan (3D surface)**

* Plot $z=f(x,y)$ di ruang 3D. Lebih sulit dibaca daripada kontur untuk menangkap posisi ekstrem, tetapi memberi gambaran bentuk lengkap.

**Tip membaca kontur**

* Garis kontur rapat  gradien besar (fungsi naik/turun cepat).
* Minimum: kontur berbentuk cincin tertutup dengan nilai menurun ke tengah.
* Saddle: kontur menunjukkan pola silang (naik di satu arah, turun di arah lain).



# 5) Pengali Lagrange (Lagrange multipliers)  untuk kendala *equality*

**Masalah dasar**
Minimalkan $f(\mathbf{x})$ s.t. $h(\mathbf{x})=0$ (bisa banyak kendala).

**Lagrangian**

$$
\mathcal{L}(\mathbf{x},\boldsymbol\lambda) = f(\mathbf{x}) + \boldsymbol\lambda^\top h(\mathbf{x}).
$$

**Kondisi stasioner**

$$
\nabla_x \mathcal{L}(\mathbf{x}^*,\boldsymbol\lambda^*) = 0,\quad h(\mathbf{x}^*)=0.
$$

Artinya gradien $f$ di titik optimum adalah kombinasi linear dari gradien kendala.

**Contoh cepat (langkah demi langkah)**
Min $f(x,y)=x^2+y^2$ s.t. $x+y=1$.

1. $\mathcal{L}=x^2+y^2+\lambda(x+y-1)$.
2. Stasioner: $2x+\lambda=0,\;2y+\lambda=0$.
3. Dari kedua: $x=y$. Dengan kendala $x+y=1$  $x=y=0.5$.
4. Jadi solusi optimum $(0.5,0.5)$.

**Interpretasi **

* Besarnya  adalah sensitivitas nilai optimum terhadap perubahan RHS kendala (seberapa banyak nilai optimum berubah kalau batas kendala digeser sedikit).



# 6) KKT (KarushKuhnTucker)  generalisasi untuk kendala *inequality*

**Masalah standar**
Min $f(\mathbf{x})$ s.t. $h_i(\mathbf{x})=0$ (equality), $g_j(\mathbf{x})\le0$ (inequality).

**Kondisi KKT (ringkas)**

1. **Stasionaritas**: $\nabla f(\mathbf{x}^*) + \sum_i \lambda_i \nabla h_i(\mathbf{x}^*) + \sum_j \mu_j \nabla g_j(\mathbf{x}^*) = 0.$
2. **Primal feasibility**: $h_i(\mathbf{x}^*)=0,\; g_j(\mathbf{x}^*)\le 0.$
3. **Dual feasibility**: $\mu_j \ge 0$ untuk semua j (multipliers untuk kendala tak-berat harus non-negatif).
4. **Complementary slackness**: $\mu_j\, g_j(\mathbf{x}^*) = 0$ untuk setiap j (artinya: untuk setiap kendala, **entah** kendala aktif $(g_j=0)$ dan $\mu_j\ge0$ atau kendala tidak aktif $(g_j<0)$ dan $\mu_j=0$).

**Contoh singkat**
Min $f(x)=x^2$ s.t. $x\ge1$. Tulis $g(x)=1-x\le0$.

* Lagrangian: $L = x^2 + \mu(1-x)$.
* Stasioner: $2x - \mu = 0$  $\mu = 2x$.
* Jika kendala tidak aktif (x>1) maka $\mu=0$  x=0 (kontradiksi). Jadi kendala aktif: x=1. Dari stasioner $\mu=2\ge0$. Solusi: $x^*=1$.

**Catatan penting**

* Untuk masalah **konveks** (f cembung, g\_j convex, h\_i linear), KKT adalah **sifat perlu & cukup** untuk optimalitas. Untuk non-konveks hanya perlu syarat (dengan asumsi regularitas).



# 7) Gradient Descent (metode iteratif dasar)

**Ide utama**
Bergerak turun sepanjang arah negatif gradien untuk mengecilkan fungsi:

$$
\mathbf{x}_{k+1} = \mathbf{x}_k - \alpha_k \,\nabla f(\mathbf{x}_k)
$$

dengan $\alpha_k>0$ (learning rate / step size).

**Sifat & tips**

* Pilih $\alpha$ terlalu besar  bisa divergen; terlalu kecil  konvergen sangat lambat.
* Untuk fungsi konveks-lipschitz gradien, descent konvergen ke minimum global (dengan pengaturan $\alpha$ tepat).
* Mudah diimplementasikan dan murah per iterasi (hanya gradien), tapi lambat mendekati optimum (terutama untuk kondisi buruk / ill-conditioned problems).

**Contoh angka**
$f=x^2+y^2$, ambil $\mathbf{x}_0=(1,1)$, $\nabla f=(2,2)$, $\alpha=0.1$.
$\mathbf{x}_1=(1,1)-0.1(2,2)=(0.8,0.8)$.

**Implementasi (pseudo-MATLAB/Scilab)**

```matlab
x = [1;1];
alpha = 0.1;
for k=1:100
  g = [2*x(1); 2*x(2)];
  x = x - alpha*g;
end
```



# 8) Newton (metode Newton multivariat)

**Formula**

$$
\mathbf{x}_{k+1} = \mathbf{x}_k - H_f(\mathbf{x}_k)^{-1}\,\nabla f(\mathbf{x}_k).
$$

**Kelebihan**

* Konvergensi **kuadratik** dekat optimum: sangat cepat bila dekat.
* Untuk fungsi kuadratik $f(\mathbf{x}) = \tfrac12\mathbf{x}^\top Q \mathbf{x} + b^\top x + c$, Newton menyelesaikan **dalam satu langkah** (jika mulai dari sembarang titik) karena $\nabla f = Qx+b$ dan $H=Q$ konstan  solusi langsung.

**Kekurangan**

* Perlu menghitung dan menyelesaikan sistem linear dengan Hessian (biaya komputasi $O(n^3)$ naive), dan Hessian harus invertible/positif definite untuk menjamin langkah menuju minimum.
* Jika Hessian tidak positif definite, perlu modifikasi (damped Newton, dogleg, add damping term).

**Contoh cepat**
$f=x^2+y^2$: Hessian = $2I$. Newton: $x_{k+1} = x_k - (2I)^{-1}(2x_k) = 0$. Satu iterasi selesai.

**Praktis**

* Biasanya dipakai kalau dimensi kecilsedang dan kita punya informasi Hessian; untuk dimensi besar sering pakai varian Quasi-Newton (BFGS) yang memperkirakan inverse Hessian.



# 9) Bagaimana semuanya terhubung (intuisi visual)

* Pada **kontur**, titik optimum terlihat sebagai pusat kontur melingkar (minimum) atau puncak (maksimum).
* **Gradient** pada tiap titik menunjuk tegak lurus ke kontur menuju nilai yang lebih tinggi. Gradient descent bergerak menuruni kontur mengikuti panah negatif gradien.
* **Hessian** memberi tahu apakah area itu cekung (bisa jadi minimum) atau ada arah naik/turun campur (saddle).
* **Lagrange / KKT** muncul bila ada batasan: optimum sering berada di titik di mana kontur fungsi tujuan menyentuh batas kendala  gradien fungsi tujuan sebanding dengan gradien kendala (Lagrange), atau untuk inequality, berlaku kondisi slackness (KKT).



# Ringkasan cepat (cheat-sheet)

* Turunan parsial: $\partial f/\partial x_i$.
* Gradient: $\nabla f$.
* Hessian: $H_f$. Pos. definite  minimum lokal.
* Contour: level sets, jarak garis  gradien.
* Permukaan: visual 3D $z=f(x,y)$.
* Lagrange: gunakan bila ada kendala equality; selesaikan $\nabla_x\mathcal{L}=0, h=0$.
* KKT: stationarity, primal/dual feasibility, complementary slackness untuk kendala inequality.
* Gradient Descent: $\mathbf{x}\leftarrow\mathbf{x}-\alpha\nabla f$. Sederhana & murah per iterasi.
* Newton: $\mathbf{x}\leftarrow\mathbf{x}-H^{-1}\nabla f$. Cepat (kuadratik), mahal (Hessian).
