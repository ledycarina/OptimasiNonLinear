# Metode Newton

Metode **Newton** (atau juga dikenal sebagai **Metode Newton-Raphson** dalam konteks pencarian akar) adalah metode iteratif yang kuat dalam mata kuliah Optimasi Non-Linear untuk mencari titik kritis (potensial minimum, maksimum, atau titik pelana) dari suatu fungsi. Secara umum, dalam optimasi tanpa kendala (Unconstrained Optimization), tujuannya adalah meminimalkan fungsi $f(\mathbf{x})$, di mana $\mathbf{x} \in \mathbb{R}^n$ adalah vektor variabel.

##  Konsep Matematis

Tujuan utama dalam optimasi adalah mencari titik $\mathbf{x}^*$ di mana gradien fungsi sama dengan nol, yaitu:
$$\nabla f(\mathbf{x}^*) = \mathbf{0}$$
Titik ini disebut **titik stasioner** atau **titik kritis**. Metode Newton beroperasi dengan mendekati fungsi tujuan $f(\mathbf{x})$ pada iterasi ke-$k$ dengan ekspansi **Taylor orde kedua** di sekitar titik iterasi saat ini $\mathbf{x}_k$.

### Ekspansi Taylor Orde Kedua

Ekspansi Taylor orde kedua dari $f(\mathbf{x})$ di sekitar $\mathbf{x}_k$ adalah:
$$f(\mathbf{x}) \approx q(\mathbf{x}) = f(\mathbf{x}_k) + \nabla f(\mathbf{x}_k)^T (\mathbf{x} - \mathbf{x}_k) + \frac{1}{2} (\mathbf{x} - \mathbf{x}_k)^T \mathbf{H}(\mathbf{x}_k) (\mathbf{x} - \mathbf{x}_k)$$
di mana:
* $\nabla f(\mathbf{x}_k)$ adalah **Gradien** (vektor turunan parsial pertama) $f$ di $\mathbf{x}_k$.
* $\mathbf{H}(\mathbf{x}_k)$ adalah **Matriks Hessian** (matriks turunan parsial kedua) $f$ di $\mathbf{x}_k$.

### Mencari Arah Newton (Newton Direction)

Metode Newton bertujuan mencari titik iterasi berikutnya $\mathbf{x}_{k+1}$ yang meminimalkan hampiran kuadratik $q(\mathbf{x})$. Misalkan $\mathbf{p}_k = \mathbf{x} - \mathbf{x}_k$ adalah langkah pergeseran (Newton Step) yang dicari. Kita ingin meminimalkan $q(\mathbf{x}_k + \mathbf{p}_k)$. Untuk mencari minimum dari $q(\mathbf{x}_k + \mathbf{p}_k)$ terhadap $\mathbf{p}_k$, kita harus menyetarakan gradien dari $q(\mathbf{x}_k + \mathbf{p}_k)$ terhadap $\mathbf{p}_k$ menjadi nol:

$$\nabla q(\mathbf{x}_k + \mathbf{p}_k) = \mathbf{0}$$
Gradien dari $q(\mathbf{x}_k + \mathbf{p}_k)$ terhadap $\mathbf{p}_k$ adalah:
$$\nabla q(\mathbf{x}_k + \mathbf{p}_k) = \nabla f(\mathbf{x}_k) + \mathbf{H}(\mathbf{x}_k) \mathbf{p}_k$$

Menyetarakan dengan nol menghasilkan sistem persamaan linear:
$$\nabla f(\mathbf{x}_k) + \mathbf{H}(\mathbf{x}_k) \mathbf{p}_k = \mathbf{0}$$

Dengan mengasumsikan Matriks **Hessian $\mathbf{H}(\mathbf{x}_k)$ tidak singular** (memiliki invers), kita dapat menyelesaikan $\mathbf{p}_k$:
$$\mathbf{p}_k = - \mathbf{H}(\mathbf{x}_k)^{-1} \nabla f(\mathbf{x}_k)$$
$\mathbf{p}_k$ inilah yang disebut **Arah Newton** (Newton Direction).

### Rumus Iterasi Metode Newton

Titik iterasi berikutnya $\mathbf{x}_{k+1}$ diperoleh dengan menambahkan $\mathbf{p}_k$ ke $\mathbf{x}_k$:
$$\mathbf{x}_{k+1} = \mathbf{x}_k + \mathbf{p}_k$$
Atau, secara lengkap:
$$\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{H}(\mathbf{x}_k)^{-1} \nabla f(\mathbf{x}_k)$$
Proses ini diulang (iterasi) sampai kondisi konvergensi terpenuhi, misalnya ketika $||\nabla f(\mathbf{x}_{k+1})||$ sangat kecil atau $|| \mathbf{x}_{k+1} - \mathbf{x}_k ||$ sangat kecil.

***

##  Penjelasan Singkat dan Mudah Dipahami

Bayangkan Anda sedang berada di bukit (permukaan fungsi $f(\mathbf{x})$) dan ingin mencari titik terendah (minimum).

1.  **Iterasi Awal ($\mathbf{x}_k$):** Anda berdiri di suatu titik awal $\mathbf{x}_0$.
2.  **Mendekati Permukaan:** Di titik Anda saat ini ($\mathbf{x}_k$), Anda menghampiri permukaan bukit lokal dengan kurva berbentuk parabola (fungsi kuadratik $q(\mathbf{x})$).
    * **Gradien ($\nabla f(\mathbf{x}_k)$):** Memberi tahu Anda arah "paling curam" bukit.
    * **Hessian ($\mathbf{H}(\mathbf{x}_k)$):** Memberi tahu Anda kelengkungan atau bentuk lokal bukit (apakah parabola terbuka ke atas/minimum, ke bawah/maksimum, atau memiliki titik pelana).
3.  **Menemukan Titik Terendah Parabola:** Anda kemudian menghitung titik terendah dari parabola hampiran tersebut. Titik terendah inilah yang menjadi langkah pergeseran $\mathbf{p}_k$ Anda.
4.  **Langkah Selanjutnya ($\mathbf{x}_{k+1}$):** Anda melangkah dari $\mathbf{x}_k$ sejauh $\mathbf{p}_k$ untuk mencapai titik baru $\mathbf{x}_{k+1} = \mathbf{x}_k + \mathbf{p}_k$.
5.  **Ulangi:** Anda mengulangi proses ini (dari langkah 2) sampai langkah pergeseran menjadi sangat kecil, yang berarti Anda sudah sangat dekat dengan titik kritis (minimum lokal).

### Keunggulan Metode Newton

Metode Newton memiliki laju konvergensi **kuadratik** di dekat solusi. Ini berarti bahwa jika tebakan awal Anda sudah cukup dekat dengan solusi, jumlah digit akurat Anda akan berlipat ganda pada setiap iterasi. Ini menjadikannya sangat cepat dibandingkan metode lain seperti **Gradient Descent** (yang memiliki konvergensi linear) yang hanya menggunakan turunan pertama (Gradien).

### Kekurangan Metode Newton

1.  **Membutuhkan Turunan Kedua (Hessian):** Perlu menghitung matriks Hessian, yang bisa sangat mahal secara komputasi dan rumit secara analitis, terutama untuk fungsi dengan banyak variabel ($n$ besar, karena Hessian berukuran $n \times n$).
2.  **Invers Hessian:** Harus menghitung invers dari matriks Hessian pada setiap iterasi, yang juga mahal komputasinya.
3.  **Hessian Tidak Positif Definit:** Jika Hessian $\mathbf{H}(\mathbf{x}_k)$ tidak **positif definit** (positive definite) di $\mathbf{x}_k$, arah Newton $\mathbf{p}_k$ mungkin bukan arah penurunan (descent direction), dan iterasi bisa divergen atau konvergen ke titik maksimum atau pelana.

Untuk mengatasi kekurangan komputasi dan masalah indefinit, metode yang lebih praktis seperti **Quasi-Newton Methods** (misalnya BFGS) dan **Levenberg-Marquardt** sering digunakan dalam praktik, yang mengestimasi (mendekati) Hessian atau invers Hessian.


# Komponen-Komponen dari Metode Newton

Komponen utama dari Metode Newton untuk Optimasi Non-Linear adalah **Gradien (Turunan Pertama)**, **Matriks Hessian (Turunan Kedua)**, dan **Langkah Newton (Newton Step)**. Ketiga komponen ini bekerja bersama dalam proses iteratif untuk menemukan titik optimum.

Berikut penjelasan detail dan terperinci untuk setiap komponen utama, berdasarkan rumus iterasi:
$$\mathbf{x}_{k+1} = \mathbf{x}_k + \mathbf{p}_k$$
di mana **Langkah Newton** $\mathbf{p}_k$ didefinisikan sebagai solusi dari sistem persamaan linear:
$$\mathbf{H}(\mathbf{x}_k) \mathbf{p}_k = - \nabla f(\mathbf{x}_k)$$
atau, jika $\mathbf{H}(\mathbf{x}_k)$ non-singular:
$$\mathbf{p}_k = - \mathbf{H}(\mathbf{x}_k)^{-1} \nabla f(\mathbf{x}_k)$$

***

## 1. Gradien ($\nabla f(\mathbf{x}_k)$)

**Gradien** adalah vektor turunan parsial pertama dari fungsi tujuan $f(\mathbf{x})$.

* **Definisi:** Untuk fungsi $f(\mathbf{x}): \mathbb{R}^n \to \mathbb{R}$, gradien di titik $\mathbf{x}_k$ adalah vektor kolom berdimensi $n$:
    $$\nabla f(\mathbf{x}_k) = \begin{pmatrix} \frac{\partial f}{\partial x_1}(\mathbf{x}_k) \\ \frac{\partial f}{\partial x_2}(\mathbf{x}_k) \\ \vdots \\ \frac{\partial f}{\partial x_n}(\mathbf{x}_k) \end{pmatrix}$$
* **Peran dalam Optimasi:** Gradien menunjukkan **arah peningkatan fungsi yang paling curam** (Steepest Ascent).
    * Dalam konteks Metode Newton, gradien memberikan informasi tentang **seberapa jauh titik saat ini dari optimum**. Ketika $\nabla f(\mathbf{x}) = \mathbf{0}$, maka $\mathbf{x}$ adalah titik kritis.
    * Gradien berfungsi sebagai **vektor sisi kanan** dalam sistem persamaan untuk mencari Langkah Newton.

***

## 2. Matriks Hessian ($\mathbf{H}(\mathbf{x}_k)$)

**Matriks Hessian** adalah matriks simetris dari turunan parsial kedua dari fungsi tujuan $f(\mathbf{x})$.

* **Definisi:** Hessian di titik $\mathbf{x}_k$ adalah matriks $n \times n$:
    $$\mathbf{H}(\mathbf{x}_k) = \nabla^2 f(\mathbf{x}_k) = \begin{pmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\ \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_n} \\ \vdots & \vdots & \ddots & \vdots \\ \frac{\partial^2 f}{\partial x_n \partial x_1} & \frac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n^2} \end{pmatrix}$$
* **Peran dalam Optimasi:** Hessian memberikan informasi tentang **kelengkungan (curvatura) lokal** dari fungsi $f(\mathbf{x})$. Ini sangat penting untuk membedakan antara minimum, maksimum, dan titik pelana (saddle point).
    * Untuk optimasi (pencarian minimum), idealnya $\mathbf{H}(\mathbf{x}_k)$ harus **positif definit** (memastikan arah $\mathbf{p}_k$ adalah arah penurunan).
    * Dalam konteks Metode Newton, Hessian berfungsi sebagai **matriks koefisien** dalam sistem persamaan untuk mencari Langkah Newton.

***

## 3. Langkah Newton ($\mathbf{p}_k$)

**Langkah Newton** adalah vektor pergeseran dari titik saat ini $\mathbf{x}_k$ ke titik iterasi berikutnya $\mathbf{x}_{k+1}$.

* **Definisi:** $\mathbf{p}_k$ adalah solusi unik dari **Persamaan Newton** (Newton Equation): $\mathbf{H}(\mathbf{x}_k) \mathbf{p}_k = - \nabla f(\mathbf{x}_k)$.
* **Peran dalam Optimasi:** $\mathbf{p}_k$ adalah langkah yang akan membawa $\mathbf{x}_k$ ke titik minimum dari aproksimasi kuadratik lokal fungsi $f(\mathbf{x})$.
    * Langkah ini secara inheren mengandung informasi tentang **magnitudo (besaran)** dan **arah** yang optimal untuk bergerak, karena ia menggabungkan informasi kemiringan (gradien) dan kelengkungan (Hessian).
    * Metode Newton standar menetapkan $\mathbf{p}_k$ sebagai **arah** dan **panjang langkah** ($\alpha_k=1$).

***

## 4. Persamaan Newton

**Persamaan Newton** adalah sistem persamaan linear yang harus diselesaikan pada setiap iterasi.

* **Bentuk:** $\mathbf{H}(\mathbf{x}_k) \mathbf{p}_k = - \nabla f(\mathbf{x}_k)$
* **Proses Komputasi:** Secara praktis, untuk $n$ besar, kita **TIDAK** menghitung invers Hessian $\mathbf{H}(\mathbf{x}_k)^{-1}$. Sebaliknya, kita menyelesaikan sistem persamaan linear ini untuk mencari $\mathbf{p}_k$ menggunakan metode seperti dekomposisi **Cholesky** (jika Hessian positif definit) atau dekomposisi **LU**. Hal ini lebih stabil dan efisien secara numerik.

***

## 5. Titik Iterasi Baru ($\mathbf{x}_{k+1}$)

Titik iterasi yang diperbarui dihasilkan dari penambahan Langkah Newton ke titik saat ini.

* **Rumus:** $\mathbf{x}_{k+1} = \mathbf{x}_k + \mathbf{p}_k$
* **Iterasi:** Proses ini diulang, di mana $\mathbf{x}_{k+1}$ menjadi $\mathbf{x}_{\text{baru}}$ dan digunakan untuk menghitung Gradien dan Hessian yang baru pada iterasi berikutnya, sampai kondisi konvergensi tercapai.

**Ringkasan Alur Kerja:**

1.  **Mulai** dari tebakan awal $\mathbf{x}_0$.
2.  Hitung **Gradien** $\nabla f(\mathbf{x}_k)$ dan **Hessian** $\mathbf{H}(\mathbf{x}_k)$.
3.  Selesaikan **Persamaan Newton** untuk mendapatkan **Langkah Newton** $\mathbf{p}_k$.
4.  Hitung titik baru $\mathbf{x}_{k+1} = \mathbf{x}_k + \mathbf{p}_k$.
5.  **Ulangi** sampai kriteria berhenti (misalnya, $|| \nabla f(\mathbf{x}_{k+1}) || < \epsilon$) terpenuhi.

# Contoh dari setiap Komponen-Komponen dalam Metode Newton

Untuk menjelaskan setiap komponen Metode Newton, saya akan menggunakan contoh fungsi sederhana (yang biasanya dipelajari dalam optimasi 1D) dan kemudian menggeneralisasinya ke 2D.

Misalkan kita ingin mencari nilai minimum dari fungsi tujuan non-linear 2-dimensi (dua variabel):
$$f(x_1, x_2) = x_1^2 + 2x_2^2 + 2x_1 x_2 - 4x_1 - 6x_2 + 5$$
Kita akan menghitung komponen-komponennya pada titik iterasi awal $\mathbf{x}_k = \mathbf{x}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$.

***

## 1. Gradien ($\nabla f(\mathbf{x}_0)$)

Gradien adalah vektor turunan parsial pertama. Ia menunjukkan arah tercuram dari fungsi.

### Perhitungan Matematis
Hitung turunan parsial pertama terhadap $x_1$ dan $x_2$:
$$\frac{\partial f}{\partial x_1} = \frac{\partial}{\partial x_1} (x_1^2 + 2x_2^2 + 2x_1 x_2 - 4x_1 - 6x_2 + 5) = 2x_1 + 2x_2 - 4$$
$$\frac{\partial f}{\partial x_2} = \frac{\partial}{\partial x_2} (x_1^2 + 2x_2^2 + 2x_1 x_2 - 4x_1 - 6x_2 + 5) = 4x_2 + 2x_1 - 6$$

Vektor gradien adalah:
$$\nabla f(\mathbf{x}) = \begin{pmatrix} 2x_1 + 2x_2 - 4 \\ 2x_1 + 4x_2 - 6 \end{pmatrix}$$

### Contoh pada Titik Awal
Substitusikan titik awal $\mathbf{x}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$:
$$\nabla f(\mathbf{x}_0) = \begin{pmatrix} 2(0) + 2(0) - 4 \\ 2(0) + 4(0) - 6 \end{pmatrix} = \begin{pmatrix} -4 \\ -6 \end{pmatrix}$$

**Interpretasi:** Pada $\mathbf{x}_0$, gradien menunjukkan vektor $\begin{pmatrix} -4 \\ -6 \end{pmatrix}$. Karena kita mencari **minimum**, gradien negatif berarti kita harus bergerak ke arah **positif** (searah dengan $-\nabla f(\mathbf{x}_0)$) untuk mengurangi nilai fungsi.

***

## 2. Matriks Hessian ($\mathbf{H}(\mathbf{x}_0)$)

Hessian adalah matriks turunan parsial kedua. Ia mendeskripsikan kelengkungan fungsi.

### Perhitungan Matematis
Hitung semua turunan parsial kedua:
1.  **Diagonal (Turunan Kedua Murni):**
    $$h_{11} = \frac{\partial^2 f}{\partial x_1^2} = \frac{\partial}{\partial x_1} (2x_1 + 2x_2 - 4) = 2$$
    $$h_{22} = \frac{\partial^2 f}{\partial x_2^2} = \frac{\partial}{\partial x_2} (2x_1 + 4x_2 - 6) = 4$$
2.  **Off-Diagonal (Turunan Kedua Campuran):**
    $$h_{12} = \frac{\partial^2 f}{\partial x_1 \partial x_2} = \frac{\partial}{\partial x_1} (2x_1 + 4x_2 - 6) = 2$$
    $$h_{21} = \frac{\partial^2 f}{\partial x_2 \partial x_1} = \frac{\partial}{\partial x_2} (2x_1 + 2x_2 - 4) = 2$$
    *(Sesuai Teorema Young, $h_{12} = h_{21}$ karena fungsi kita mulus, menjadikan Hessian simetris).*

Matriks Hessian adalah:
$$\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 2 & 2 \\ 2 & 4 \end{pmatrix}$$

### Contoh pada Titik Awal
Karena fungsi tujuan kita adalah fungsi kuadratik, Hessian-nya adalah konstanta, sehingga:
$$\mathbf{H}(\mathbf{x}_0) = \begin{pmatrix} 2 & 2 \\ 2 & 4 \end{pmatrix}$$

**Interpretasi:** Kelengkungan fungsi adalah konstan. Untuk menentukan apakah ini minimum, kita perlu memeriksa **definiteness** dari $\mathbf{H}$.
* Determinan minor utama pertama: $2 > 0$.
* Determinan minor utama kedua (determinan matriks $\mathbf{H}$): $2(4) - 2(2) = 8 - 4 = 4 > 0$.
Karena kedua minor utama positif, $\mathbf{H}$ adalah **positif definit**. Ini memastikan bahwa $\mathbf{p}_k$ yang kita hitung akan menjadi **arah penurunan** yang baik.

***

## 3. Persamaan Newton dan Langkah Newton ($\mathbf{p}_k$)

Langkah Newton $\mathbf{p}_k$ diperoleh dengan menyelesaikan sistem persamaan linear $\mathbf{H}(\mathbf{x}_k) \mathbf{p}_k = - \nabla f(\mathbf{x}_k)$.

### Komponen Persamaan
$$\begin{pmatrix} 2 & 2 \\ 2 & 4 \end{pmatrix} \begin{pmatrix} p_1 \\ p_2 \end{pmatrix} = - \begin{pmatrix} -4 \\ -6 \end{pmatrix}$$
$$\begin{pmatrix} 2 & 2 \\ 2 & 4 \end{pmatrix} \begin{pmatrix} p_1 \\ p_2 \end{pmatrix} = \begin{pmatrix} 4 \\ 6 \end{pmatrix}$$

### Solusi untuk $\mathbf{p}_0$ (Langkah Newton)
Kita dapat menyelesaikan sistem persamaan ini (misalnya menggunakan invers matriks, karena matriks 2x2 mudah dibalik).

1.  **Hitung Invers Hessian:**
    $$\mathbf{H}(\mathbf{x}_0)^{-1} = \frac{1}{4} \begin{pmatrix} 4 & -2 \\ -2 & 2 \end{pmatrix} = \begin{pmatrix} 1 & -1/2 \\ -1/2 & 1/2 \end{pmatrix}$$
2.  **Hitung Langkah Newton:**
    $$\mathbf{p}_0 = - \mathbf{H}(\mathbf{x}_0)^{-1} \nabla f(\mathbf{x}_0) = \begin{pmatrix} 1 & -1/2 \\ -1/2 & 1/2 \end{pmatrix} \begin{pmatrix} 4 \\ 6 \end{pmatrix}$$
    $$\mathbf{p}_0 = \begin{pmatrix} 1(4) - 1/2(6) \\ -1/2(4) + 1/2(6) \end{pmatrix} = \begin{pmatrix} 4 - 3 \\ -2 + 3 \end{pmatrix} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$$

**Interpretasi:** Vektor $\mathbf{p}_0 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$ adalah Langkah Newton. Ini berarti kita harus bergerak 1 unit ke arah $x_1$ positif dan 1 unit ke arah $x_2$ positif.

***

## 4. Titik Iterasi Baru ($\mathbf{x}_{1}$)

Titik iterasi berikutnya adalah $\mathbf{x}_1 = \mathbf{x}_0 + \mathbf{p}_0$.

### Perhitungan
$$\mathbf{x}_{1} = \begin{pmatrix} 0 \\ 0 \end{pmatrix} + \begin{pmatrix} 1 \\ 1 \end{pmatrix} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$$

### Hasil
Titik iterasi baru adalah $\mathbf{x}_1 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$.

Karena fungsi tujuan kita adalah fungsi kuadratik (yang dapat dimodelkan secara sempurna oleh ekspansi Taylor orde kedua), **Metode Newton konvergen dalam satu iterasi**. Mari kita periksa gradien di $\mathbf{x}_1$:
$$\nabla f(\mathbf{x}_1) = \begin{pmatrix} 2(1) + 2(1) - 4 \\ 2(1) + 4(1) - 6 \end{pmatrix} = \begin{pmatrix} 4 - 4 \\ 6 - 6 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$$
Karena gradiennya nol, $\mathbf{x}_1 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$ adalah titik minimum dari fungsi tersebut.

# Hubungan Metode Newton dengan Metode Newton-Raphson

 **Metode Newton** dalam optimasi multivariabel (fungsi $f(\mathbf{x})$ dengan $\mathbf{x} \in \mathbb{R}^n$) memiliki hubungan fundamental dengan **Metode Newton-Raphson** yang Anda pelajari di optimasi satu dimensi (1D) atau pencarian akar (root-finding).

Hubungannya adalah bahwa **Metode Newton untuk Optimasi** adalah **generalisasi dimensi tinggi (multivariabel) dari Metode Newton-Raphson 1D yang diterapkan pada turunan pertama (gradien) dari fungsi tujuan.**

***

## 1. Konteks Pencarian Akar (Root-Finding) â€” Metode Newton-Raphson (1D)

Di aljabar numerik, Metode Newton-Raphson digunakan untuk mencari akar dari persamaan $g(x)=0$, di mana $x$ adalah skalar (1D).

### Formula Iterasi (1D)
Metode ini menggunakan ekspansi Taylor orde pertama dari $g(x)$ di sekitar $x_k$ dan menyetarakannya dengan nol:
$$g(x) \approx g(x_k) + g'(x_k) (x - x_k) = 0$$
Menyelesaikan untuk $x = x_{k+1}$ menghasilkan:
$$x_{k+1} = x_k - \frac{g(x_k)}{g'(x_k)}$$

***

## 2. Aplikasi ke Optimasi (1D)

Dalam optimasi tanpa kendala, tujuan kita adalah meminimalkan (atau memaksimalkan) fungsi $f(x)$. Titik optimum $x^*$ adalah titik kritis yang memenuhi kondisi orde pertama:
$$f'(x^*) = 0$$

### Menghubungkan $f(x)$ dengan $g(x)$
Jika kita mendefinisikan fungsi $g(x)$ sebagai **turunan pertama** dari $f(x)$:
$$g(x) = f'(x)$$
Maka, mencari titik optimum $x^*$ dari $f(x)$ sama dengan mencari akar $x^*$ dari $g(x)$, yaitu, mencari $x^*$ sehingga $g(x^*) = 0$.

### Menerapkan Newton-Raphson ke $g(x)$
Sekarang, kita terapkan rumus iterasi Newton-Raphson dari bagian 1 ke fungsi $g(x)$:

* **Fungsi:** $g(x) = f'(x)$
* **Turunan Fungsi:** $g'(x) = f''(x)$

Substitusi ke dalam rumus Newton-Raphson (1D):
$$x_{k+1} = x_k - \frac{g(x_k)}{g'(x_k)}$$
$$x_{k+1} = x_k - \frac{f'(x_k)}{f''(x_k)}$$

Inilah **Metode Newton untuk Optimasi 1D**.

***

## 3. Generalisasi ke Optimasi Multivariabel (n-Dimensi)

Metode Newton untuk optimasi non-linear yang Anda pelajari $(\mathbf{x} \in \mathbb{R}^n)$ adalah generalisasi langsung dari persamaan di atas.

| Konsep Optimasi 1D ($x$) | Konsep Optimasi $n$-Dimensi ($\mathbf{x}$) |
| :--- | :--- |
| **Turunan Pertama** $f'(x)$ | **Gradien** $\nabla f(\mathbf{x})$ (vektor) |
| **Turunan Kedua** $f''(x)$ | **Hessian** $\mathbf{H}(\mathbf{x})$ (matriks) |
| **Pembagian** $\frac{1}{f''(x)}$ | **Invers Matriks** $\mathbf{H}(\mathbf{x})^{-1}$ |

### Formula Iterasi (n-Dimensi)
Rumus iterasi Newton untuk Optimasi $n$-dimensi:
$$\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{H}(\mathbf{x}_k)^{-1} \nabla f(\mathbf{x}_k)$$

### Kesimpulan Hubungan

Metode Newton dalam Optimasi Non-Linear (multivariabel) adalah **generalisasi matriks dan vektor** dari Metode Newton-Raphson (1D) yang digunakan untuk mencari akar $\nabla f(\mathbf{x}) = \mathbf{0}$.

| Metode | Tujuan | Formula General |
| :--- | :--- | :--- |
| **Newton-Raphson (1D)** | Cari akar $g(x)=0$ | $x_{k+1} = x_k - g(x_k)/g'(x_k)$ |
| **Newton (Optimasi)** | Cari optimum $f(x)$ | $\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{H}(\mathbf{x}_k)^{-1} \nabla f(\mathbf{x}_k)$ |

Dalam kedua kasus, metode ini mengandalkan **hampiran kuadratik lokal** fungsi untuk menentukan langkah selanjutnya.

## Contoh 1: Termudah (Optimasi 1D, Kuadratik)

Ini adalah kasus paling dasar di mana Metode Newton konvergen dalam satu iterasi karena fungsi tersebut sudah merupakan aproksimasi kuadratik yang sempurna.

**Tujuan:** Minimalkan fungsi $f(x) = x^2 - 4x + 7$.
**Tebakan Awal:** $x_0 = 0$.

### Komponen
1.  **Turunan Pertama (Gradien 1D):**
    $$f'(x) = 2x - 4$$
2.  **Turunan Kedua (Hessian 1D):**
    $$f''(x) = 2$$
3.  **Langkah Newton ($p_k$):**
    $$p_k = - \frac{f'(x_k)}{f''(x_k)}$$

### Iterasi
**Iterasi $k=0$:**
* $f'(x_0) = f'(0) = 2(0) - 4 = -4$
* $f''(x_0) = f''(0) = 2$
* $p_0 = - \frac{-4}{2} = 2$
* **Titik Baru:** $x_1 = x_0 + p_0 = 0 + 2 = 2$

**Pemeriksaan:**
$f'(x_1) = f'(2) = 2(2) - 4 = 0$. **Optimum ditemukan.**

**Kesimpulan:** Minimum dari $f(x)$ terletak di $x^* = 2$.

## Contoh 2: Menengah (Optimasi 1D, Non-Kuadratik, Multi-Iterasi)

Fungsi non-kuadratik membutuhkan beberapa iterasi karena aproksimasi kuadratik lokal hanya berlaku di sekitar $x_k$.

**Tujuan:** Minimalkan fungsi $f(x) = x^4 - 3x^3 + 2$.
**Tebakan Awal:** $x_0 = 2$.
**Toleransi:** $\epsilon = 10^{-4}$ (untuk $\left|f'(x)\right|$).

### Komponen
1.  **Turunan Pertama:** $f'(x) = 4x^3 - 9x^2$
2.  **Turunan Kedua:** $f''(x) = 12x^2 - 18x$

### Iterasi
**Iterasi $k=0$ ($x_0 = 2$):**
* $f'(2) = 4(2)^3 - 9(2)^2 = 32 - 36 = -4$
* $f''(2) = 12(2)^2 - 18(2) = 48 - 36 = 12$
* $p_0 = - \frac{-4}{12} = \frac{1}{3} \approx 0.3333$
* **Titik Baru:** $x_1 = 2 + 0.3333 = 2.3333$

**Iterasi $k=1$ ($x_1 = 2.3333$):**
* $f'(x_1) = 4(2.3333)^3 - 9(2.3333)^2 \approx 0.7407$
* $f''(x_1) = 12(2.3333)^2 - 18(2.3333) \approx 19.3331$
* $p_1 = - \frac{0.7407}{19.3331} \approx -0.0383$
* **Titik Baru:** $x_2 = 2.3333 - 0.0383 = 2.2950$

**Iterasi $k=2$ ($x_2 = 2.2950$):**
* $f'(x_2) \approx 0.0068$ (Sudah mendekati 0)
* $f''(x_2) \approx 18.7303$
* $p_2 = - \frac{0.0068}{18.7303} \approx -0.0004$
* **Titik Baru:** $x_3 = 2.2950 - 0.0004 = 2.2946$

**Kesimpulan:** Optimum lokal adalah $x^* \approx 2.2946$.

## Contoh 3: Lebih Sulit (Optimasi 2D, Kuadratik Sederhana)

Ini adalah generalisasi ke multivariabel. Kuadratik berarti konvergensi dalam satu langkah.

**Tujuan:** Minimalkan fungsi $f(\mathbf{x}) = x_1^2 + 2x_2^2 - 4x_1 + 8x_2 + 10$.
**Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$.

### Komponen
1.  **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} 2x_1 - 4 \\ 4x_2 + 8 \end{pmatrix}$
2.  **Hessian:** $\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix}$ (Konstan)

### Iterasi
**Iterasi $k=0$:**
* $\nabla f(\mathbf{x}_0) = \begin{pmatrix} 2(0) - 4 \\ 4(0) + 8 \end{pmatrix} = \begin{pmatrix} -4 \\ 8 \end{pmatrix}$
* $\mathbf{H}(\mathbf{x}_0) = \begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix}$
* **Persamaan Newton:** $\mathbf{H}(\mathbf{x}_0) \mathbf{p}_0 = - \nabla f(\mathbf{x}_0)$
    $$\begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix} \mathbf{p}_0 = \begin{pmatrix} 4 \\ -8 \end{pmatrix}$$
* **Langkah Newton ($\mathbf{p}_0$):**
    $$\mathbf{p}_0 = \begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix}^{-1} \begin{pmatrix} 4 \\ -8 \end{pmatrix} = \begin{pmatrix} 1/2 & 0 \\ 0 & 1/4 \end{pmatrix} \begin{pmatrix} 4 \\ -8 \end{pmatrix} = \begin{pmatrix} 2 \\ -2 \end{pmatrix}$$
* **Titik Baru:** $\mathbf{x}_1 = \mathbf{x}_0 + \mathbf{p}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix} + \begin{pmatrix} 2 \\ -2 \end{pmatrix} = \begin{pmatrix} 2 \\ -2 \end{pmatrix}$

**Pemeriksaan:**
$\nabla f(\mathbf{x}_1) = \begin{pmatrix} 2(2) - 4 \\ 4(-2) + 8 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$. **Optimum ditemukan.**

**Kesimpulan:** Minimum dari $f(\mathbf{x})$ terletak di $\mathbf{x}^* = \begin{pmatrix} 2 \\ -2 \end{pmatrix}$.

## Contoh 4: Sulit (Optimasi 2D, Non-Kuadratik, Multi-Iterasi)

Ini adalah kasus nyata yang umum, di mana fungsi non-kuadratik membutuhkan perhitungan gradien dan Hessian baru pada setiap iterasi.

**Tujuan:** Minimalkan fungsi Rosenbrock 2D yang disederhanakan: $f(\mathbf{x}) = (1-x_1)^2 + (x_2 - x_1^2)^2$.
**Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$.

### Komponen
1.  **Gradien:**
    $$\nabla f(\mathbf{x}) = \begin{pmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \end{pmatrix} = \begin{pmatrix} -2(1-x_1) - 4x_1(x_2 - x_1^2) \\ 2(x_2 - x_1^2) \end{pmatrix}$$
2.  **Hessian:**
    $$\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 2 - 4x_2 + 12x_1^2 & -4x_1 \\ -4x_1 & 2 \end{pmatrix}$$

### Iterasi
**Iterasi $k=0$ ($\mathbf{x}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$):**
* $\nabla f(\mathbf{x}_0) = \begin{pmatrix} -2(1) - 0 \\ 2(0) \end{pmatrix} = \begin{pmatrix} -2 \\ 0 \end{pmatrix}$
* $\mathbf{H}(\mathbf{x}_0) = \begin{pmatrix} 2 - 0 + 0 & 0 \\ 0 & 2 \end{pmatrix} = \begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix}$
* **Langkah Newton ($\mathbf{p}_0$):**
    $$\mathbf{p}_0 = - \mathbf{H}(\mathbf{x}_0)^{-1} \nabla f(\mathbf{x}_0) = - \begin{pmatrix} 1/2 & 0 \\ 0 & 1/2 \end{pmatrix} \begin{pmatrix} -2 \\ 0 \end{pmatrix} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$
* **Titik Baru:** $\mathbf{x}_1 = \mathbf{x}_0 + \mathbf{p}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix} + \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$

**Iterasi $k=1$ ($\mathbf{x}_1 = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$):**
* $\nabla f(\mathbf{x}_1) = \begin{pmatrix} -2(1-1) - 4(1)(0 - 1) \\ 2(0 - 1) \end{pmatrix} = \begin{pmatrix} 4 \\ -2 \end{pmatrix}$
* $\mathbf{H}(\mathbf{x}_1) = \begin{pmatrix} 2 - 0 + 12 & -4 \\ -4 & 2 \end{pmatrix} = \begin{pmatrix} 14 & -4 \\ -4 & 2 \end{pmatrix}$
* **Langkah Newton ($\mathbf{p}_1$):** (Dengan menyelesaikan sistem persamaan $\mathbf{H}\mathbf{p} = -\nabla f$)
    $$\mathbf{p}_1 \approx \begin{pmatrix} -0.333 \\ -0.333 \end{pmatrix}$$
* **Titik Baru:** $\mathbf{x}_2 = \mathbf{x}_1 + \mathbf{p}_1 = \begin{pmatrix} 1 \\ 0 \end{pmatrix} + \begin{pmatrix} -0.333 \\ -0.333 \end{pmatrix} = \begin{pmatrix} 0.667 \\ -0.333 \end{pmatrix}$

**Catatan:** Minimum sebenarnya adalah $\mathbf{x}^* = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$. Metode ini akan terus beriterasi hingga mencapai solusi tersebut dengan cepat.

## Contoh 5: Paling Sulit (Metode Newton dengan *Line Search*)

Dalam praktiknya, langkah Newton yang diperoleh mungkin terlalu besar dan menyebabkan fungsi naik (divergen), terutama jauh dari optimum. Untuk itu, ditambahkan **teknik Line Search** untuk membatasi panjang langkah $\alpha_k$.

**Rumus Iterasi Modifikasi (Damped Newton):**
$$\mathbf{x}_{k+1} = \mathbf{x}_k + \alpha_k \mathbf{p}_k, \quad \text{di mana } \alpha_k \in (0, 1] \text{ dipilih melalui line search.}$$

**Tujuan:** Minimalkan fungsi dari Contoh 4: $f(\mathbf{x}) = (1-x_1)^2 + (x_2 - x_1^2)^2$.
**Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} -1.2 \\ 1.0 \end{pmatrix}$. (Titik awal yang jauh)

### Iterasi 0 (sama seperti Contoh 4, tetapi dengan $\mathbf{x}_0$ yang berbeda)
1.  **Hitung Gradien dan Hessian di $\mathbf{x}_0$:**
    * $\nabla f(\mathbf{x}_0) \approx \begin{pmatrix} -5.28 \\ 0.8 \end{pmatrix}$
    * $\mathbf{H}(\mathbf{x}_0) \approx \begin{pmatrix} 14.88 & 4.8 \\ 4.8 & 2 \end{pmatrix}$
2.  **Hitung Arah Newton $\mathbf{p}_0$:**
    * $\mathbf{p}_0 = - \mathbf{H}(\mathbf{x}_0)^{-1} \nabla f(\mathbf{x}_0) \approx \begin{pmatrix} 0.584 \\ -1.748 \end{pmatrix}$

3.  **Langkah $p_0$ Penuh (tanpa Line Search):**
    * $\mathbf{x}_{\text{full}} = \mathbf{x}_0 + \mathbf{p}_0 \approx \begin{pmatrix} -1.2 \\ 1.0 \end{pmatrix} + \begin{pmatrix} 0.584 \\ -1.748 \end{pmatrix} = \begin{pmatrix} -0.616 \\ -0.748 \end{pmatrix}$
    * $f(\mathbf{x}_0) \approx 4.84$
    * $f(\mathbf{x}_{\text{full}}) \approx 1.87$ (Fungsi *turun*, langkah penuh $\alpha=1$ mungkin baik)

4.  **Terapkan Line Search (misalnya, dengan Aturan Armijo/Backtracking):**
    * Coba $\alpha = 1$: $f(\mathbf{x}_{\text{full}})$ menunjukkan penurunan yang signifikan.
    * **Pilih:** $\alpha_0 = 1$.
    * **Titik Baru:** $\mathbf{x}_1 = \mathbf{x}_0 + 1 \cdot \mathbf{p}_0 \approx \begin{pmatrix} -0.616 \\ -0.748 \end{pmatrix}$

### Iterasi 1
Proses diulangi: Hitung $\nabla f(\mathbf{x}_1)$, $\mathbf{H}(\mathbf{x}_1)$, hitung $\mathbf{p}_1$, dan cari $\alpha_1$ yang paling baik.

**Penting:** Line Search memastikan bahwa $f(\mathbf{x}_{k+1}) < f(\mathbf{x}_k)$ pada setiap langkah, yang merupakan jaminan konvergensi global (walaupun kecepatannya melambat jika $\alpha_k < 1$). Metode ini lebih kuat dan stabil.

Konvergensi **Metode Newton** itu mudahnya bisa dibilang sangat **cepat ðŸš€**, tapi metodenya agak **pemilih** soal titik awal.

Pembahasan konvergensi berfokus pada dua hal utama: seberapa cepat ia bekerja, dan kapan ia dijamin berhasil.

***

## 1. Kecepatan: Konvergensi Kuadratik (Super Cepat!)

Metode Newton terkenal karena memiliki **konvergensi kuadratik**.

### Apa Artinya?
Bayangkan akurasi solusi Anda. Konvergensi kuadratik berarti **jumlah digit akurat yang Anda miliki akan berlipat ganda pada setiap langkah iterasi**.

* Jika pada iterasi ke-5 Anda memiliki 4 digit yang benar (misalnya, 1.3333...),
* Maka pada iterasi ke-6 Anda bisa mendapatkan sekitar 8 digit yang benar (misalnya, 1.33333333...),
* Dan pada iterasi ke-7, Anda bisa mendapatkan 16 digit yang benar!

Inilah sebabnya Metode Newton sering menjadi pilihan utama ketika kecepatan sangat dibutuhkan.

***

## 2. Jaminan Keberhasilan: Konvergensi Lokal

Meskipun super cepat, Metode Newton **hanya dijamin berhasil (konvergen) jika Anda memulainya dari tempat yang tepat**. Ini disebut **konvergensi lokal**.

### Syarat Wajib agar Berhasil

1.  **Titik Awal Harus Dekat (Pemilih):** Titik awal ($\mathbf{x}_0$) harus sudah berada di "lingkungan" yang cukup dekat dengan solusi akhir ($\mathbf{x}^*$). Jika terlalu jauh, Metode Newton bisa **menjauhi solusi (divergen)** atau malah menemukan maksimum, bukan minimum.
2.  **Permukaan "Mulus" dan Cekung:** Fungsi yang Anda minimalkan ($f(\mathbf{x})$) harus:
    * Sangat mulus (turunan pertama dan kedua ada dan kontinu).
    * Memiliki kelengkungan positif di sekitar solusi. Secara teknis, ini berarti Matriks **Hessian** (turunan kedua) harus **positif definit** di titik optimum. Ini memastikan Anda benar-benar menuju minimum, bukan titik pelana (saddle point) atau maksimum.

***

## 3. Modifikasi untuk Stabilitas

Karena Metode Newton sangat "pemilih" soal titik awal dan bisa divergen, dalam praktik nyata, ia sering dimodifikasi untuk meningkatkan **konvergensi global** (bisa dimulai dari mana saja).

### Metode Newton Terhambat (Damped Newton Method)
Modifikasi paling umum adalah menambahkan **panjang langkah ($\alpha$)** yang disebut **Line Search**:

$$\mathbf{x}_{k+1} = \mathbf{x}_k + \alpha_k \cdot \mathbf{p}_k$$

* **$\mathbf{p}_k$** adalah arah cepat yang diberikan oleh Newton.
* **$\alpha_k$** adalah angka antara 0 dan 1 yang dipilih pada setiap langkah.

**Fungsi $\alpha$:**
Tujuan $\alpha$ adalah **memastikan** bahwa setiap langkah **pasti menurunkan nilai fungsi** ($f(\mathbf{x}_{k+1}) < f(\mathbf{x}_k)$).

* Jika langkah Newton penuh ($\alpha=1$) menyebabkan nilai fungsi naik, Line Search akan mencoba $\alpha=1/2$, kemudian $\alpha=1/4$, dan seterusnya, hingga ditemukan langkah yang aman.

Dengan Line Search, konvergensi global terjamin. Namun, di awal iterasi, saat $\alpha < 1$, kecepatannya akan sedikit menurun (menjadi konvergensi linear), tetapi akan kembali menjadi **kuadratik** setelah Anda mendekati solusi ($\alpha \to 1$).



### Ilustrasi Metode Newton (Python)

Kita akan meminimalkan fungsi Rosenbrock sederhana:
$$f(x_1, x_2) = (1-x_1)^2 + (x_2 - x_1^2)^2$$
Optimum global berada di $\mathbf{x}^* = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$ dengan $f(\mathbf{x}^*)=0$.


````python
import numpy as np

# --- 1. DEFENISI KOMPONEN ANALITIK ---

def f(x):
    """Fungsi Tujuan (Rosenbrock Sederhana)"""
    x1, x2 = x
    return (1 - x1)**2 + (x2 - x1**2)**2

def gradient(x):
    """Vektor Gradien (Turunan Pertama)"""
    x1, x2 = x
    df_dx1 = -2 * (1 - x1) - 4 * x1 * (x2 - x1**2)
    df_dx2 = 2 * (x2 - x1**2)
    return np.array([df_dx1, df_dx2])

def hessian(x):
    """Matriks Hessian (Turunan Kedua)"""
    x1, x2 = x
    # H[0, 0] = d^2f / dx1^2
    H_11 = 2 - 4 * x2 + 12 * x1**2
    # H[1, 1] = d^2f / dx2^2
    H_22 = 2
    # H[0, 1] = H[1, 0] = d^2f / dx1 dx2
    H_12 = -4 * x1
    return np.array([
        [H_11, H_12],
        [H_12, H_22]
    ])

# --- 2. ALGORITMA METODE NEWTON ---

def newton_optimization(x_start, max_iter=10, tolerance=1e-5):
    x_k = np.array(x_start, dtype=float)

    print(f"{'Iter':<5}{'x1':<10}{'x2':<10}{'||Grad||':<10}{'f(x)':<10}")
    print("-" * 45)

    for k in range(max_iter):

        # 1. Hitung Gradien (g_k) dan Hessian (H_k) pada titik saat ini x_k
        g_k = gradient(x_k)
        H_k = hessian(x_k)

        # Norm Gradien untuk kriteria berhenti
        grad_norm = np.linalg.norm(g_k)
        f_val = f(x_k)

        print(f"{k:<5}{x_k[0]:<10.4f}{x_k[1]:<10.4f}{grad_norm:<10.4e}{f_val:<10.4e}")

        # Kriteria Berhenti: Jika Gradien sudah mendekati nol
        if grad_norm < tolerance:
            print("\nKonvergensi Tercapai (Gradien mendekati nol).")
            break

        # 2. Selesaikan Persamaan Newton: H_k * p_k = -g_k
        try:
            # Menggunakan np.linalg.solve() lebih stabil daripada mencari invers (np.linalg.inv())
            p_k = np.linalg.solve(H_k, -g_k)
        except np.linalg.LinAlgError:
            print(f"\nPeringatan: Hessian singular pada iterasi {k}. Menghentikan proses.")
            break

        # 3. Hitung Titik Iterasi Baru: x_{k+1} = x_k + p_k
        # Untuk metode Newton murni, panjang langkah alpha adalah 1 (p_k = alpha * p_k)
        alpha = 1.0
        x_k = x_k + alpha * p_k

        if k == max_iter - 1:
            print("\nBatasan Iterasi Maksimum Tercapai.")

    return x_k

# --- 3. EKSEKUSI SIMULASI ---

# Mulai dari titik yang jauh dari optimum
x_initial = (-1.2, 1.0)
print(f"Memulai Metode Newton dengan x0 = {x_initial}\n")

x_optimal = newton_optimization(x_initial)

print("\n--- Hasil Akhir ---")
print(f"Titik Optimum yang Ditemukan (x*): {x_optimal}")
print(f"Nilai Fungsi Minimum f(x*): {f(x_optimal):.6e}")
````

## 1\. Contoh Sangat Mudah: Kuadratik Cekung (Konvergensi 1 Langkah)

Ini adalah kasus paling sederhana: fungsi kuadratik dengan sumbu yang sejajar. Metode Newton menemukan minimum dalam satu langkah, karena aproksimasi kuadratik lokal adalah fungsi itu sendiri.

### 1.1 Masalah dan Analisis

  * **Fungsi Tujuan:** $f(\mathbf{x}) = x_1^2 + x_2^2 + 5$.
      * Permukaannya adalah mangkuk cekung yang sempurna.
  * **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} 2x_1 \\ 2x_2 \end{pmatrix}$
  * **Hessian:** $\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix}$ (Konstan, Positif Definit)
  * **Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$.
  * **Langkah Newton:** $\mathbf{p}_0 = - \mathbf{H}^{-1} \nabla f(\mathbf{x}_0) = - \begin{pmatrix} 1/2 & 0 \\ 0 & 1/2 \end{pmatrix} \begin{pmatrix} 4 \\ 2 \end{pmatrix} = \begin{pmatrix} -2 \\ -1 \end{pmatrix}$.
  * **Hasil:** $\mathbf{x}_1 = \mathbf{x}_0 + \mathbf{p}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ (Minimum ditemukan dalam satu langkah).

### 1.2 Ilustrasi 3D (Python)




````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f1(x): return x[0]**2 + x[1]**2 + 5
path1 = np.array([[2.0, 1.0], [0.0, 0.0]])
path1_z = np.array([f1(p) for p in path1])

x1_range, x2_range = np.linspace(-3, 3, 50), np.linspace(-3, 3, 50)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f1([X1, X2])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='coolwarm', alpha=0.7)
ax.plot(path1[:, 0], path1[:, 1], path1_z, color='k', marker='o', markersize=5, linewidth=4)
ax.set_title('C1: Kuadratik Cekung (1 Langkah)')
plt.show()
````



## 2\. Contoh Mudah: Kuadratik Miring (Perubahan Arah Sumbu)

Meskipun masih kuadratik dan konvergen dalam 1 langkah, fungsi ini memiliki istilah silang ($x_1x_2$), yang membuat Hessian memiliki komponen *off-diagonal*. Ini menunjukkan bagaimana $\mathbf{H}^{-1}$ secara otomatis menyesuaikan langkah agar sejajar dengan arah sumbu baru.

### 2.1 Masalah dan Analisis

  * **Fungsi Tujuan:** $f(\mathbf{x}) = 2x_1^2 + x_2^2 + x_1x_2 - 6x_1 - 4x_2$.
      * Permukaannya adalah mangkuk cekung, tetapi **terpelintir** (sumbu cekungannya miring).
  * **Hessian:** $\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 4 & 1 \\ 1 & 2 \end{pmatrix}$ (Konstan).
  * **Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$.
  * **Langkah Newton:** $\mathbf{p}_0 = \begin{pmatrix} 1.6 \\ 1.2 \end{pmatrix}$ (Dihitung dari $\mathbf{H}^{-1} \nabla f$).

````python
# Import yang diperlukan sudah ada
def f2(x): return 2*x[0]**2 + x[1]**2 + x[0]*x[1] - 6*x[0] - 4*x[1]
path2 = np.array([[0.0, 0.0], [1.6, 1.2]])
path2_z = np.array([f2(p) for p in path2])

x1_range, x2_range = np.linspace(-1, 3, 50), np.linspace(-1, 3, 50)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f2([X1, X2])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.7)
ax.plot(path2[:, 0], path2[:, 1], path2_z, color='k', marker='o', markersize=5, linewidth=4)
ax.set_title('C2: Kuadratik Miring (Hessian Non-Diagonal)')
plt.show()
````

## 3\. Contoh Menengah: Non-Kuadratik Sederhana (Multi-Iterasi)

Ini adalah contoh pertama yang membutuhkan beberapa iterasi karena Hessian berubah pada setiap langkah. Namun, fungsinya cukup mulus, sehingga konvergensi masih sangat cepat (kuadratik).

### 3.1 Masalah dan Analisis

  * **Fungsi Tujuan:** $f(\mathbf{x}) = x_1^4 + x_2^4 + x_1^2 + x_2^2 + x_1x_2$.
      * Permukaan yang sangat mulus dan cembung, tetapi turunan orde tingginya non-nol.
  * **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} 4x_1^3 + 2x_1 + x_2 \\ 4x_2^3 + 2x_2 + x_1 \end{pmatrix}$
  * **Hessian:** $\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 12x_1^2 + 2 & 1 \\ 1 & 12x_2^2 + 2 \end{pmatrix}$ (Bergantung pada $\mathbf{x}$)
  * **Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$.
  * **Hasil:** Konvergensi kuadratik dalam 3-4 langkah menuju $\mathbf{x}^* = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$.

### 3.2 Ilustrasi 3D (Python)

````python
def f3(x): return x[0]**4 + x[1]**4 + x[0]**2 + x[1]**2 + x[0]*x[1]
def gradient_complex(x):
    x1, x2 = x; return np.array([4*x1**3 + 2*x1 + x2, 4*x2**3 + 2*x2 + x1])
def hessian_complex(x):
    x1, x2 = x; return np.array([[12*x1**2 + 2, 1.0], [1.0, 12*x2**2 + 2]])

def newton_tracker(x_start):
    x_k = np.array(x_start); path = [x_k.copy()]
    for _ in range(5):
        g_k = gradient_complex(x_k); H_k = hessian_complex(x_k)
        if np.linalg.norm(g_k) < 1e-4: break
        p_k = np.linalg.solve(H_k, -g_k)
        x_k = x_k + p_k
        path.append(x_k.copy())
    return np.array(path)

path3 = newton_tracker([1.0, 1.0])
path3_z = np.array([f3(p) for p in path3])

x1_range, x2_range = np.linspace(-1.5, 1.5, 50), np.linspace(-1.5, 1.5, 50)
X1, X2 = np.meshgrid(x1_range, x2_range); Z = f3([X1, X2])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='plasma', alpha=0.7)
ax.plot(path3[:, 0], path3[:, 1], path3_z, color='w', marker='o', markersize=5, linewidth=3)
ax.set_title('C3: Non-Kuadratik Sederhana (Konvergensi Cepat)')
plt.show()
````

### 4.1 Masalah dan Analisis

  * **Fungsi Tujuan:** $f(\mathbf{x}) = x_1^2 - x_2^2$ (Fungsi Pelana/Hyperbolic Paraboloid).
      * Minimum sepanjang $x_2$ (cekung ke atas), Maksimum sepanjang $x_1$ (cekung ke bawah).
  * **Hessian:** $\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 2 & 0 \\ 0 & -2 \end{pmatrix}$ (Indefinit).
  * **Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$.
  * **Langkah Newton:** $\mathbf{p}_0 = - \mathbf{H}^{-1} \nabla f(\mathbf{x}_0) = - \begin{pmatrix} 1/2 & 0 \\ 0 & -1/2 \end{pmatrix} \begin{pmatrix} 2 \\ -2 \end{pmatrix} = \begin{pmatrix} -1 \\ -1 \end{pmatrix}$.
  * **Hasil:** $\mathbf{x}_1 = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$. Titik pelana ditemukan dalam 1 langkah\! (Metode Newton untuk *root-finding* $\nabla f(\mathbf{x})=0$).

### 4.2 Ilustrasi 3D (Python)

````python
def f4(x): return x[0]**2 - x[1]**2 # Fungsi Pelana
path4 = np.array([[1.0, 1.0], [0.0, 0.0]])
path4_z = np.array([f4(p) for p in path4])

x1_range, x2_range = np.linspace(-2, 2, 50), np.linspace(-2, 2, 50)
X1, X2 = np.meshgrid(x1_range, x2_range); Z = f4([X1, X2])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='bwr', alpha=0.8) # Warna Bipolar untuk Saddle
ax.plot(path4[:, 0], path4[:, 1], path4_z, color='k', marker='o', markersize=5, linewidth=4)
ax.set_title('C4: Titik Pelana (Hessian Indefinit)')
plt.show()
````


## 5\. Contoh Paling Sulit: Lembah Rosenbrock (Konvergensi Lambat di Lembah Sempit)

Ini adalah studi kasus klasik. Fungsi ini memiliki minimum global di dasar lembah yang sangat sempit (banana shape). Metode Newton di sini menunjukkan konvergensi kuadratik hanya setelah ia mencapai lembah, tetapi dapat mengambil langkah besar yang tidak akurat di awal.

### 5.1 Masalah dan Analisis

  * **Fungsi Tujuan:** $f(\mathbf{x}) = 100(x_2 - x_1^2)^2 + (1 - x_1)^2$ (Fungsi Rosenbrock standar).
  * **Hessian:** Sangat non-linear dan berubah-ubah; menjadi **singular** (buruk) di sepanjang lembah.
  * **Tebakan Awal:** $\mathbf{x}_0 = \begin{pmatrix} -1 \\ 1 \end{pmatrix}$.
  * **Hasil:** Diperlukan sekitar 15-20 langkah untuk konvergensi penuh ke $\mathbf{x}^* = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$. Jalur iterasi menunjukkan langkah besar (kadang terlalu besar) diikuti oleh langkah-langkah kecil dan cepat.

### 5.2 Ilustrasi 3D (Python)

````python
def f5(x): return 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2
def gradient_rosen(x): # Didefinisikan di atas
    x1, x2 = x
    return np.array([-400*x1*(x2 - x1**2) - 2*(1 - x1), 200*(x2 - x1**2)])
def hessian_rosen(x): # Didefinisikan di atas
    x1, x2 = x
    H_11 = 1200*x1**2 - 400*x2 + 2
    H_12 = -400*x1
    return np.array([[H_11, H_12], [H_12, 200]])

def newton_tracker_rosen(x_start):
    x_k = np.array(x_start); path = [x_k.copy()]
    for _ in range(8): # Batasi 8 langkah untuk visualisasi yang jelas
        g_k = gradient_rosen(x_k); H_k = hessian_rosen(x_k)
        if np.linalg.norm(g_k) < 1e-4: break
        try:
            p_k = np.linalg.solve(H_k, -g_k)
            x_k = x_k + p_k
            path.append(x_k.copy())
        except np.linalg.LinAlgError: break
    return np.array(path)

path5 = newton_tracker_rosen([-1.0, 1.0])
path5_z = np.array([f5(p) for p in path5])

x1_range, x2_range = np.linspace(-1.5, 1.5, 100), np.linspace(-0.5, 2.0, 100)
X1, X2 = np.meshgrid(x1_range, x2_range); Z = f5([X1, X2])
# Batasi nilai Z tinggi agar lembah terlihat jelas
Z[Z > 100] = 100

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='terrain', alpha=0.6)
ax.plot(path5[:, 0], path5[:, 1], path5_z, color='r', marker='o', markersize=5, linewidth=3)
ax.set_title('C5: Fungsi Rosenbrock (Konvergensi Terhambat)')
ax.view_init(elev=20, azim=-140)
plt.show()
````


## Ilustrasi Paling Indah: Fungsi Lembah Rosenbrock (3D) ðŸŒ¹

### Masalah dan Komponen

Fungsi Rosenbrock adalah standar *benchmark* dalam optimasi yang sering disebut "lembah pisang." Minimum global berada di $\mathbf{x}^* = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$.

  * **Fungsi Tujuan:** $f(\mathbf{x}) = 100(x_2 - x_1^2)^2 + (1 - x_1)^2$
  * **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} -400x_1(x_2 - x_1^2) - 2(1 - x_1) \\ 200(x_2 - x_1^2) \end{pmatrix}$
  * **Hessian:** $\mathbf{H}(\mathbf{x}) = \begin{pmatrix} 1200x_1^2 - 400x_2 + 2 & -400x_1 \\ -400x_1 & 200 \end{pmatrix}$

### Kode Python dan Visualisasi

Kode di bawah akan menghitung 10 langkah iterasi dari titik awal yang jauh dan memvisualisasikan jalur tersebut di atas permukaan 3D yang diwarnai dengan gradien yang menarik.





````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. Definisi Fungsi dan Turunannya (Komponen Newton) ---

def f_rosen(x):
    """Fungsi Rosenbrock 2D"""
    return 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2

def gradient_rosen(x):
    """Gradien/Turunan Pertama"""
    x1, x2 = x
    df_dx1 = -400*x1*(x2 - x1**2) - 2*(1 - x1)
    df_dx2 = 200*(x2 - x1**2)
    return np.array([df_dx1, df_dx2])

def hessian_rosen(x):
    """Matriks Hessian/Turunan Kedua"""
    x1, x2 = x
    H_11 = 1200*x1**2 - 400*x2 + 2
    H_12 = -400*x1
    return np.array([[H_11, H_12], [H_12, 200]])

# --- 2. Algoritma Optimasi dan Pelacakan Jalur ---

def newton_tracker_rosen(x_start):
    x_k = np.array(x_start, dtype=float)
    path = [x_k.copy()]

    # Maksimal 10 langkah untuk menjaga visualisasi tetap jelas
    for _ in range(10):
        g_k = gradient_rosen(x_k); H_k = hessian_rosen(x_k)
        if np.linalg.norm(g_k) < 1e-6: break

        try:
            # Hitung Langkah Newton: p_k = H_k^-1 * (-g_k)
            p_k = np.linalg.solve(H_k, -g_k)

            # Langkah Terhambat (Damped Newton) untuk stabilitas di Rosenbrock
            # Kita gunakan alpha=0.5 di awal untuk menghindari divergensi, lalu alpha=1
            alpha = 0.5 if np.linalg.norm(x_k - np.array([1, 1])) > 1.5 else 1.0

            x_k = x_k + alpha * p_k
            path.append(x_k.copy())
        except np.linalg.LinAlgError:
            break # Hentikan jika Hessian singular

    return np.array(path)

# --- 3. EKSEKUSI dan VISUALISASI 3D ---

x_initial = np.array([-1.0, 1.0])
path = newton_tracker_rosen(x_initial)

# Persiapan Grid untuk Plotting
x1_range, x2_range = np.linspace(-1.2, 1.5, 100), np.linspace(-0.5, 2.0, 100)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f_rosen([X1, X2])

# Batasi nilai Z tinggi (agar lembah terlihat jelas, memotong "dinding")
Z[Z > 100] = 100

# Data Jalur Iterasi (X, Y, Z)
path_x1, path_x2 = path[:, 0], path[:, 1]
path_z = np.array([f_rosen(p) for p in path])

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot Permukaan Fungsi dengan cmap yang menarik
ax.plot_surface(X1, X2, Z, cmap='terrain', alpha=0.6, rstride=5, cstride=5)

# Plot Jalur Iterasi Newton (Garis tebal dan kontras)
ax.plot(path_x1, path_x2, path_z, color='red', marker='o', markersize=6,
        label='Jalur Iterasi Newton', linewidth=3.5, zorder=3)

# Tandai Titik Awal dan Akhir
ax.scatter(path_x1[0], path_x2[0], path_z[0], color='black', s=120, label='Titik Awal (x0)', zorder=4)
ax.scatter(path_x1[-1], path_x2[-1], path_z[-1], color='gold', marker='*', s=400, label='Optimum (x*)', zorder=5)

# Pengaturan Label dan Judul
ax.set_xlabel('$x_1$', fontsize=12)
ax.set_ylabel('$x_2$', fontsize=12)
ax.set_zlabel('$f(x)$ (Tinggi Permukaan)', fontsize=12)
ax.set_title("Metode Newton: Navigasi Lembah Rosenbrock", fontsize=16, pad=20)
ax.legend(loc='upper right')
ax.view_init(elev=20, azim=-140) # Sudut pandang yang optimal untuk melihat lembah
plt.show()
````

1.  **Topografi Sulit (Permukaan Hijau):** Permukaan `terrain` menampilkan lembah yang sangat sempit dan landai (di mana minimum berada), dikelilingi oleh "dinding" yang curam. Inilah yang membuat optimasi sulit.
2.  **Jalur Iterasi (Garis Merah):** Jalur merah yang tebal dan kontras secara dramatis menunjukkan bagaimana Newton *melompat* ke dasar lembah, kemudian **berliku-liku** di sepanjang dasar lembah.

## 1. Aspek Teoritis Mendalam (Derivasi dan Interpretasi)

Materi yang baik harus menjelaskan **mengapa** Metode Newton bekerja.

### A. Derivasi Lengkap
Jelaskan asal muasal rumus iterasi ($\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{H}^{-1} \nabla f(\mathbf{x}_k)$) dari **Ekspansi Deret Taylor Orde Kedua** di sekitar $\mathbf{x}_k$. Tunjukkan bagaimana meminimalkan aproksimasi kuadratik ini menghasilkan langkah Newton.

### B. Interpretasi Geometris
Jelaskan bahwa:
* $\nabla f(\mathbf{x}_k)$ (Gradien) menentukan arah paling curam (lereng).
* $\mathbf{H}(\mathbf{x}_k)$ (Hessian) menentukan **kelengkungan** permukaan.
* Langkah Newton ($\mathbf{p}_k$) adalah vektor yang membawa kita langsung ke dasar **parabola lokal** yang mengaproksimasi fungsi $f(\mathbf{x})$. Ini membedakannya dari Metode Gradient Descent yang hanya menggunakan informasi lereng.

***

## 2. Keterbatasan dan Kegagalan (Kondisi Nyata)

Materi harus membahas kapan metode ini **gagal** di dunia nyata.

### A. Kondisi Singularitas Hessian
Apa yang terjadi jika Matriks Hessian ($\mathbf{H}$) menjadi **singular** (tidak memiliki invers) atau **ill-conditioned** (hampir singular, menyebabkan langkah $\mathbf{p}_k$ menjadi sangat besar dan tidak akurat)? Ini sering terjadi:
* Di dekat daerah datar.
* Pada titik-titik yang jauh dari optimum (membuat aproksimasi kuadratik buruk).

### B. Masalah Non-Konveksitas
Bahasa yang digunakan dalam optimasi: Apa yang terjadi jika $\mathbf{H}$ adalah **indefinit** (titik pelana)? Metode Newton akan menemukan titik kritis ($\nabla f=0$), tetapi **tanpa modifikasi** (seperti yang dibahas pada Contoh 4), ia tidak akan tahu apakah itu minimum, maksimum, atau titik pelana.

***

## 3. Implementasi Praktis dan Kompleksitas

Ini mencakup detail teknis yang diperlukan untuk menjalankan metode secara efisien.

### A. Biaya Komputasi
Bandingkan biaya komputasi per iterasi:
* **Gradient Descent:** Murah, hanya perlu menghitung $\nabla f$.
* **Metode Newton:** Mahal karena perlu menghitung $\mathbf{H}$ (membutuhkan $n^2$ turunan kedua) dan kemudian **menyelesaikan sistem linier** $\mathbf{H} \mathbf{p} = -\nabla f$, yang biasanya membutuhkan $O(n^3)$ operasi (misalnya, menggunakan faktorisasi LU atau Cholesky).

### B. Strategi Penyelesaian Sistem Linier
Tekankan bahwa dalam praktik, $\mathbf{p}_k = - \mathbf{H}^{-1} \nabla f$ **tidak pernah** dihitung dengan mencari invers secara eksplisit. Sebaliknya, digunakan algoritma stabil seperti **Faktorisasi Cholesky** (jika $\mathbf{H}$ positif definit) atau **Faktorisasi LU** untuk menyelesaikan sistem $\mathbf{H} \mathbf{p} = -\nabla f$.

***

## 4. Alternatif dan Hubungan dengan Metode Lain

Pembahasan yang lengkap harus meletakkan Metode Newton dalam konteks algoritma optimasi lainnya.

### A. Metode Kuasi-Newton
Ini adalah alternatif vital yang mengatasi biaya komputasi Newton murni.
* Jelaskan bahwa **Metode Quasi-Newton (QN)** menghindari perhitungan $\mathbf{H}$ dan $\mathbf{H}^{-1}$ secara eksplisit.
* Sebaliknya, ia **mengaproksimasi** invers Hessian ($\mathbf{B}_k \approx \mathbf{H}^{-1}$) menggunakan informasi Gradien dari iterasi sebelumnya.
* Sebutkan algoritma terkenal: **BFGS** (Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno) dan **DFP**.

### B. Perbandingan dengan Gradient Descent
Tabel perbandingan sederhana sangat membantu:

| Fitur | Metode Gradient Descent | Metode Newton |
| :--- | :--- | :--- |
| **Informasi Turunan** | Orde 1 ($\nabla f$) | Orde 2 ($\nabla f$ dan $\mathbf{H}$) |
| **Kecepatan Konvergensi** | Linear (Lambat) | Kuadratik (Sangat Cepat) |
| **Biaya per Iterasi** | Murah ($O(n)$) | Mahal ($O(n^3)$) |
| **Jaminan Konvergensi** | Global, Asalkan $\alpha$ benar | Lokal |

***

## 5. Algoritma Khusus (Trust Region)

Metode Line Search (Damped Newton) adalah satu cara untuk mengatasi masalah konvergensi. **Trust Region** adalah cara yang lebih canggih dan sering lebih stabil, dan ini harus disebutkan.

### Konsep Trust Region
Daripada memilih *panjang* langkah ($\alpha$) di arah $\mathbf{p}_k$, Metode Trust Region memilih *langkah* $\mathbf{s}_k$ yang meminimalkan aproksimasi kuadratik, tetapi **membatasi panjang** langkah tersebut dalam sebuah area kepercayaan (radius $\Delta_k$):

$$\min_{\|\mathbf{s}\| \leq \Delta_k} \left[ f(\mathbf{x}_k) + \nabla f(\mathbf{x}_k)^T \mathbf{s} + \frac{1}{2} \mathbf{s}^T \mathbf{H}_k \mathbf{s} \right]$$

Radius $\Delta_k$ disesuaikan pada setiap iterasi berdasarkan seberapa baik aproksimasi kuadratik tersebut mencocokkan perilaku fungsi aslinya. Trust Region seringkali merupakan algoritma yang lebih disukai di lingkungan perangkat lunak optimasi profesional.

# Metode Kuasi-Newton (Quasi-Newton Methods)

Metode Kuasi-Newton (QN), juga dikenal sebagai metode **sekant berdimensi banyak** (multidimensional secant method), adalah solusi brilian yang dirancang untuk mempertahankan **kecepatan konvergensi tinggi** Metode Newton tanpa menanggung **biaya komputasi yang mahal** dari perhitungan dan inversi Matriks Hessian ($\mathbf{H}$).

## 1. Motivasi Matematis: Masalah Metode Newton

Metode Newton murni menggunakan formula iterasi yang berasal dari aproksimasi Taylor Orde Kedua:

$$\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{H}_k^{-1} \nabla f(\mathbf{x}_k)$$

Masalah matematisnya ada pada $\mathbf{H}_k^{-1}$:

1.  **Biaya:** Menghitung elemen $\mathbf{H}_k$ membutuhkan $O(n^2)$ turunan kedua. Menyelesaikan $\mathbf{H}_k \mathbf{p}_k = -\nabla f$ membutuhkan $O(n^3)$ operasi. Ini terlalu mahal untuk masalah optimasi skala besar ($n$ besar).
2.  **Ketersediaan:** Turunan kedua mungkin sulit atau mustahil untuk dihitung secara analitis.

## 2. Inti Matematis Kuasi-Newton: Aproksimasi Hessian

Metode Kuasi-Newton mengganti Matriks Hessian ($\mathbf{H}_k$) dengan Matriks Aproksimasi ($\mathbf{B}_k$) atau Matriks Aproksimasi Invers Hessian ($\mathbf{D}_k$ atau $\mathbf{W}_k$).

### Rumus Iterasi Kuasi-Newton

Metode Kuasi-Newton menggantikan rumus Newton:

$$\mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{B}_k^{-1} \nabla f(\mathbf{x}_k) \quad \text{atau} \quad \mathbf{x}_{k+1} = \mathbf{x}_k - \mathbf{D}_k \nabla f(\mathbf{x}_k)$$

di mana:
* **$\mathbf{B}_k \approx \mathbf{H}_k$** (Aproksimasi Hessian).
* **$\mathbf{D}_k \approx \mathbf{H}_k^{-1}$** (Aproksimasi Invers Hessian).

**Keunggulan:** $\mathbf{B}_k$ atau $\mathbf{D}_k$ dihitung dengan biaya yang sangat rendah, seringkali hanya $O(n^2)$ atau bahkan $O(n)$ operasi, jauh lebih efisien daripada $O(n^3)$ Newton.

## 3. Kondisi Sekant (Sekant Condition)

Bagaimana $\mathbf{B}_k$ atau $\mathbf{D}_k$ dihitung? Mereka harus memenuhi kondisi yang disebut **Kondisi Sekant** (Secant Condition) atau **Persamaan Kuasi-Newton**.

Kondisi ini berasal dari **Teorema Nilai Rata-rata** yang diterapkan pada turunan pertama ($\nabla f$). Secara intuitif, kita menuntut agar aproksimasi Hessian (atau inversnya) bekerja seperti Hessian *asli* sepanjang langkah terakhir yang diambil.

### Definisi Vektor Langkah

Kita definisikan dua vektor yang mewakili langkah terakhir:

1.  **Langkah Posisi (Position Step):** Perubahan posisi antara iterasi $k$ dan $k-1$:
    $$\mathbf{s}_{k-1} = \mathbf{x}_k - \mathbf{x}_{k-1}$$

2.  **Langkah Gradien (Gradient Step):** Perubahan gradien antara iterasi $k$ dan $k-1$:
    $$\mathbf{y}_{k-1} = \nabla f(\mathbf{x}_k) - \nabla f(\mathbf{x}_{k-1})$$

### Persamaan Kuasi-Newton

Kita ingin agar $\mathbf{B}_k$ memenuhi:

$$\mathbf{B}_k \mathbf{s}_{k-1} = \mathbf{y}_{k-1}$$

**Interpretasi:** Ini adalah analogi dari persamaan $\mathbf{H}_k \mathbf{s}_{k-1} \approx \mathbf{y}_{k-1}$ yang berasal dari Taylor Orde Kedua. Kondisi ini memastikan bahwa $\mathbf{B}_k$ secara akurat mencerminkan kelengkungan rata-rata yang dialami fungsi sepanjang garis $\mathbf{s}_{k-1}$.

## 4. Proses Iterasi Pembaruan (Updating)

Metode Kuasi-Newton bersifat **iteratif** dan **memperbarui** aproksimasi $\mathbf{B}_{k-1}$ atau $\mathbf{D}_{k-1}$ menjadi $\mathbf{B}_k$ atau $\mathbf{D}_k$.

Pembaruan selalu mengambil bentuk *rank-one* atau *rank-two* (penambahan matriks dengan rank 1 atau 2).

### Pembaruan Rank-One vs. Rank-Two

* **Pembaruan Rank-One:** $\mathbf{B}_k = \mathbf{B}_{k-1} + \mathbf{U}_k$, di mana $\mathbf{U}_k$ adalah matriks rank 1. (Contoh: Metode DFP awal, yang kini jarang digunakan).
* **Pembaruan Rank-Two:** $\mathbf{B}_k = \mathbf{B}_{k-1} + \mathbf{U}_k + \mathbf{V}_k$, di mana $\mathbf{U}_k$ dan $\mathbf{V}_k$ adalah matriks rank 1. (Contoh: **BFGS**, yang paling dominan).

Pembaruan rank-two lebih disukai karena memungkinkan kita untuk **menjaga sifat positif definit** dari matriks aproksimasi, yang krusial untuk menjamin arah pencarian minimum (menjaga $\mathbf{p}_k$ sebagai arah penurunan).

## 5. Metode BFGS (Broyden-Fletcher-Goldfarb-Shanno)

**BFGS** adalah algoritma Kuasi-Newton yang paling sukses dan paling banyak digunakan.

Metode BFGS secara langsung memperbarui **Invers Hessian Aproksimasi** ($\mathbf{D}_k$).

### Rumus Pembaruan Invers Hessian $\mathbf{D}_k$ (BFGS)

$$\mathbf{D}_k = \left( \mathbf{I} - \frac{\mathbf{s}_{k-1} \mathbf{y}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}} \right) \mathbf{D}_{k-1} \left( \mathbf{I} - \frac{\mathbf{y}_{k-1} \mathbf{s}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}} \right) + \frac{\mathbf{s}_{k-1} \mathbf{s}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}}$$

**Kunci Matematis BFGS:**

* Pembaruan ini menjamin bahwa jika $\mathbf{D}_{k-1}$ positif definit, maka **$\mathbf{D}_k$ juga positif definit**, asalkan Kondisi Kurvatur Kuat (Strong Curvature Condition) terpenuhi, yaitu $\mathbf{y}_{k-1}^T \mathbf{s}_{k-1} > 0$.
* $\mathbf{D}_k$ memenuhi Kondisi Sekant untuk invers Hessian: $\mathbf{D}_k \mathbf{y}_{k-1} = \mathbf{s}_{k-1}$.

## Ringkasan Keunggulan Matematis

Metode Kuasi-Newton berhasil menggabungkan yang terbaik dari dua dunia:
1.  **Kecepatan:** Mencapai **konvergensi superlinear** (hampir secepat kuadratik Newton) tanpa memerlukan turunan kedua.
2.  **Efisiensi:** Biaya per iterasi didominasi oleh perkalian matriks-vektor, yaitu $O(n^2)$, jauh lebih murah daripada $O(n^3)$ Metode Newton murni.

# rumus-rumus utama yang digunakan dalam pendekatan Kuasi-Newton

Metode Kuasi-Newton (QN) menggunakan serangkaian rumus yang berpusat pada pembaruan **Matriks Aproksimasi Hessian ($\mathbf{B}_k$)** atau **Matriks Aproksimasi Invers Hessian ($\mathbf{D}_k$)** tanpa menghitung turunan kedua.



***

## 1. Rumus Iterasi Dasar (Langkah Pencarian)

Langkah Kuasi-Newton ($\mathbf{p}_k$) dihitung untuk menentukan arah perpindahan:
$$\mathbf{x}_{k+1} = \mathbf{x}_k + \alpha_k \mathbf{p}_k$$

1.  **Arah Kuasi-Newton ($\mathbf{p}_k$):** Dihitung dengan menggantikan invers Hessian $\mathbf{H}_k^{-1}$ dari Metode Newton dengan Matriks Aproksimasi Invers Hessian $\mathbf{D}_k$.
    $$\mathbf{p}_k = -\mathbf{D}_k \nabla f(\mathbf{x}_k)$$
    *(Catatan: $\mathbf{D}_k$ adalah $n \times n$ dan $\nabla f(\mathbf{x}_k)$ adalah Gradien $n \times 1$.)*

2.  **Panjang Langkah ($\alpha_k$):** Ditemukan melalui proses *Line Search* (Pencarian Garis) untuk memastikan nilai fungsi menurun. Biasanya $0 < \alpha_k \leq 1$.

***

## 2. Definisi Vektor Langkah

Setiap rumus pembaruan (update) didasarkan pada informasi yang dikumpulkan dari iterasi terakhir. Vektor-vektor ini harus dihitung terlebih dahulu:

1.  **Langkah Posisi (Position Step) $\mathbf{s}_{k-1}$:** Perubahan posisi dari langkah sebelumnya.
    $$\mathbf{s}_{k-1} = \mathbf{x}_k - \mathbf{x}_{k-1}$$

2.  **Langkah Gradien (Gradient Step) $\mathbf{y}_{k-1}$:** Perubahan gradien dari langkah sebelumnya.
    $$\mathbf{y}_{k-1} = \nabla f(\mathbf{x}_k) - \nabla f(\mathbf{x}_{k-1})$$

***

## 3. Kondisi Sekant (Kuasi-Newton Condition)

Matriks aproksimasi $\mathbf{B}_k$ harus memenuhi **Kondisi Sekant**, yang merupakan inti matematis dari metode Kuasi-Newton. Ini meniru sifat Hessian asli ($\mathbf{H}_k \mathbf{s}_{k-1} \approx \mathbf{y}_{k-1}$):

$$\mathbf{B}_k \mathbf{s}_{k-1} = \mathbf{y}_{k-1}$$

Persamaan ini memastikan bahwa $\mathbf{B}_k$ secara akurat mencerminkan kelengkungan rata-rata fungsi sepanjang garis $\mathbf{s}_{k-1}$.

***

## 4. Rumus Pembaruan (Update Formulas)

Rumus ini digunakan untuk menghitung $\mathbf{D}_k$ atau $\mathbf{B}_k$ dari $\mathbf{D}_{k-1}$ atau $\mathbf{B}_{k-1}$ dan vektor $\mathbf{s}_{k-1}$ serta $\mathbf{y}_{k-1}$. Rumus yang paling dominan adalah **BFGS** dan **DFP** (keduanya adalah *rank-two updates*).

### A. Rumus BFGS (Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno)

BFGS adalah yang paling populer dan paling stabil, biasanya memperbarui **Invers Hessian Aproksimasi** ($\mathbf{D}_k$).

$$\mathbf{D}_k = \left( \mathbf{I} - \frac{\mathbf{s}_{k-1} \mathbf{y}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}} \right) \mathbf{D}_{k-1} \left( \mathbf{I} - \frac{\mathbf{y}_{k-1} \mathbf{s}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}} \right) + \frac{\mathbf{s}_{k-1} \mathbf{s}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}}$$

### B. Rumus DFP (Davidonâ€“Fletcherâ€“Powell)

DFP adalah salah satu metode Kuasi-Newton tertua. Rumusnya simetris dengan BFGS, karena DFP memperbarui $\mathbf{D}_k$ sementara rumus BFGS, dalam bentuk matriks $\mathbf{B}_k$, adalah pembaruan $\mathbf{B}_k$ DFP.

$$\mathbf{D}_k = \mathbf{D}_{k-1} + \frac{\mathbf{s}_{k-1} \mathbf{s}_{k-1}^T}{\mathbf{y}_{k-1}^T \mathbf{s}_{k-1}} - \frac{(\mathbf{D}_{k-1} \mathbf{y}_{k-1}) (\mathbf{D}_{k-1} \mathbf{y}_{k-1})^T}{\mathbf{y}_{k-1}^T \mathbf{D}_{k-1} \mathbf{y}_{k-1}}$$

### C. Kondisi Kurvatur Kuat (Strong Curvature Condition)

Agar pembaruan Kuasi-Newton (terutama BFGS) dapat menjaga sifat **positif definit** dari $\mathbf{D}_k$ (yang penting untuk menjamin arah penurunan), harus dipastikan bahwa:

$$\mathbf{y}_{k-1}^T \mathbf{s}_{k-1} > 0$$

Kondisi ini umumnya dipenuhi secara otomatis jika *Line Search* yang dilakukan cukup akurat.

***

## 5. Rumus Alternatif (L-BFGS)

Untuk masalah optimasi skala sangat besar ($n \gg 1000$), di mana menyimpan matriks $\mathbf{D}_k$ berukuran $n \times n$ di memori menjadi tidak praktis, digunakan versi memori terbatas.

### L-BFGS (Limited-memory BFGS)

L-BFGS menghindari penyimpanan matriks $\mathbf{D}_k$ secara penuh. Sebaliknya, ia menyimpan hanya **$m$ pasang vektor terakhir** $(\mathbf{s}_i, \mathbf{y}_i)$, di mana $m$ biasanya kecil (misalnya, $m=5$ hingga $m=20$).

Langkah pencarian $\mathbf{p}_k$ dihitung menggunakan **algoritma dua *loop* (two-loop recursion)**, yang secara efektif mereplikasi perkalian $\mathbf{D}_k \nabla f(\mathbf{x}_k)$ tanpa matriks eksplisit, dengan biaya $O(mn)$.

$$\mathbf{p}_k = - \text{Two-Loop-Recursion}(\nabla f(\mathbf{x}_k), \{\mathbf{s}_i, \mathbf{y}_i\}_{i=k-m}^{k-1})$$

Ini adalah rumus yang paling sering digunakan dalam aplikasi modern (machine learning, komputasi skala besar).

# Komponen utama dalam Quasi Newton

Komponen utama di dalam **Metode Kuasi-Newton (Quasi-Newton Methods)** beroperasi dalam sebuah siklus iteratif yang menggabungkan kecepatan Metode Newton dengan efisiensi Metode Orde Pertama.



## 1. Titik dan Gradien Saat Ini ($\mathbf{x}_k$, $\nabla f(\mathbf{x}_k)$)

Ini adalah informasi dasar yang harus selalu tersedia pada setiap iterasi $k$.

* **Titik Posisi ($\mathbf{x}_k$):** Vektor posisi saat ini di mana evaluasi fungsi dan turunannya dilakukan.
* **Gradien ($\nabla f(\mathbf{x}_k)$):** Vektor yang menunjukkan kemiringan dan arah paling curam dari fungsi pada $\mathbf{x}_k$. Ini adalah satu-satunya informasi turunan yang **secara eksplisit dihitung** di setiap iterasi QN.

***

## 2. Matriks Aproksimasi ($\mathbf{D}_k$ atau $\mathbf{B}_k$)

Ini adalah jantung dari Metode Kuasi-Newton, yang menggantikan Matriks Hessian $\mathbf{H}_k$.

* **Matriks $\mathbf{B}_k$:** Aproksimasi Matriks Hessian **($\mathbf{B}_k \approx \mathbf{H}_k$)**.
* **Matriks $\mathbf{D}_k$:** Aproksimasi Invers Matriks Hessian **($\mathbf{D}_k \approx \mathbf{H}_k^{-1}$)**.
    * Dalam implementasi modern seperti **BFGS**, yang disimpan dan diperbarui adalah $\mathbf{D}_k$ karena memudahkan perhitungan arah $\mathbf{p}_k$.

**Fungsi:** Menyimpan informasi kelengkungan fungsi yang dikumpulkan dari riwayat gradien, memungkinkan metode untuk meniru sifat konvergensi orde kedua Newton.

***

## 3. Vektor Riwayat (Riwayat Langkah Terakhir)

Untuk memperbarui Matriks Aproksimasi ($\mathbf{D}_k$ atau $\mathbf{B}_k$), metode QN memerlukan informasi dari langkah sebelumnya ($k-1$).

* **Langkah Posisi ($\mathbf{s}_{k-1}$):** Vektor perbedaan antara posisi saat ini dan sebelumnya.
    $$\mathbf{s}_{k-1} = \mathbf{x}_k - \mathbf{x}_{k-1}$$
* **Langkah Gradien ($\mathbf{y}_{k-1}$):** Vektor perbedaan antara gradien saat ini dan sebelumnya.
    $$\mathbf{y}_{k-1} = \nabla f(\mathbf{x}_k) - \nabla f(\mathbf{x}_{k-1})$$

**Fungsi:** Kedua vektor ini menyediakan arah dan besaran perubahan Gradien yang digunakan untuk "mengajari" Matriks Aproksimasi tentang kelengkungan fungsi yang sebenarnya.

***

## 4. Rumus Pembaruan (Update Formula)

Ini adalah rumus matematis yang menerapkan **Kondisi Sekant** untuk mendapatkan matriks aproksimasi baru ($\mathbf{D}_k$) dari matriks lama ($\mathbf{D}_{k-1}$) dan vektor riwayat ($\mathbf{s}_{k-1}, \mathbf{y}_{k-1}$).

* **BFGS Update (Paling Umum):** Menggunakan pembaruan rank-dua untuk menghasilkan $\mathbf{D}_k$.
    $$\mathbf{D}_k = \left( \mathbf{I} - \frac{\mathbf{s} \mathbf{y}^T}{\mathbf{y}^T \mathbf{s}} \right) \mathbf{D}_{k-1} \left( \mathbf{I} - \frac{\mathbf{y} \mathbf{s}^T}{\mathbf{y}^T \mathbf{s}} \right) + \frac{\mathbf{s} \mathbf{s}^T}{\mathbf{y}^T \mathbf{s}}$$
    (Di mana subskrip $k-1$ disingkat untuk keterbacaan).

**Fungsi:** Memperbarui kelengkungan secara efisien ($O(n^2)$) sambil menjaga sifat-sifat yang baik (seperti positif definit, yang menjamin $\mathbf{p}_k$ adalah arah penurunan).

***

## 5. Arah Pencarian ($\mathbf{p}_k$)

Ini adalah arah vektor yang harus diikuti untuk mencapai minimum.

* **Arah Kuasi-Newton:** Dihitung dengan mengalikan Matriks Aproksimasi Invers Hessian ($\mathbf{D}_k$) dengan negatif Gradien.
    $$\mathbf{p}_k = -\mathbf{D}_k \nabla f(\mathbf{x}_k)$$

**Fungsi:** Arah ini menggabungkan informasi lereng ($\nabla f$) dengan informasi kelengkungan ($\mathbf{D}_k$) untuk menentukan jalur yang paling optimal, mirip dengan yang dilakukan Metode Newton.

***

## 6. Penentu Panjang Langkah ($\alpha_k$)

Setelah arah $\mathbf{p}_k$ ditemukan, harus ditentukan sejauh mana harus melangkah ke arah tersebut.

* **Panjang Langkah ($\alpha_k$):** Skalar yang ditemukan melalui proses **Line Search (Pencarian Garis)**.
* **Line Search:** Adalah prosedur iteratif internal (misalnya, menggunakan kondisi Armijo-Goldstein atau Wolfe) untuk mencari $\alpha_k \in (0, 1]$ yang secara signifikan mengurangi nilai fungsi.

**Fungsi:** Memastikan konvergensi global dan mencegah langkah terlalu besar yang menyebabkan divergensi atau kenaikan nilai fungsi.

***

## Siklus Iteratif QN (Runtutan)

Proses Kuasi-Newton berjalan dalam urutan sebagai berikut:

1.  **Evaluasi:** Hitung $\nabla f(\mathbf{x}_k)$.
2.  **Arah:** Hitung $\mathbf{p}_k = -\mathbf{D}_k \nabla f(\mathbf{x}_k)$.
3.  **Langkah:** Lakukan *Line Search* untuk menemukan $\alpha_k$.
4.  **Pembaruan Posisi:** Hitung posisi baru: $\mathbf{x}_{k+1} = \mathbf{x}_k + \alpha_k \mathbf{p}_k$.
5.  **Pembaruan Gradien:** Hitung $\nabla f(\mathbf{x}_{k+1})$.
6.  **Pembaruan Matriks (QN Update):** Hitung $\mathbf{s}_k$ dan $\mathbf{y}_k$, lalu gunakan **BFGS Formula** untuk memperbarui $\mathbf{D}_{k+1}$ dari $\mathbf{D}_k$.
7.  **Ulangi:** Naikkan $k$ menjadi $k+1$ dan kembali ke langkah 2.

## 1. Contoh Sangat Mudah: Metode Sekant 1D (Prekursor QN)

Ini adalah analog 1D dari Kuasi-Newton, di mana turunan kedua (Hessian 1D) diaproksimasi menggunakan metode sekant (dua titik Gradien).

### Masalah dan Rumus

* **Fungsi:** $f(x) = x^2 - 3x + 1$
* **Gradien:** $f'(x) = 2x - 3$
* **Invers Hessian (Aproksimasi):** $d_k \approx 1/f''(x)$.
* **Rumus Sekant:** $d_k = \frac{x_k - x_{k-1}}{f'(x_k) - f'(x_{k-1})} = \frac{s_{k-1}}{y_{k-1}}$
* **Iterasi:** $x_{k+1} = x_k - d_k f'(x_k)$

### Penyelesaian

| $k$ | $x_k$ | $f'(x_k)$ | $s_{k-1}$ | $y_{k-1}$ | $d_k = s/y$ | $x_{k+1}$ |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| 0 | 0.00 | -3.00 | - | - | - | - |
| 1 | 1.00 | -1.00 | $1.00$ | $2.00$ | $0.50$ | $1.50$ |
| 2 | 1.50 | 0.00 | $0.50$ | $1.00$ | $0.50$ | $\mathbf{1.50}$ |

* **Hasil:** Metode sekant (analog QN 1D) menemukan minimum analitik ($x^*=1.5$) dalam 2 langkah karena fungsinya kuadratik.

***

## 2. Contoh Mudah: Kuadratik Sederhana 2D (Logika Sekant 2D)

Kita akan menunjukkan bagaimana aproksimasi Hessian $\mathbf{B}$ harus memenuhi kondisi sekant dalam 2D pada langkah pertama.

### Masalah dan Rumus

* **Fungsi:** $f(\mathbf{x}) = x_1^2 + x_2^2$ (Minimum di $\mathbf{x}^* = \mathbf{0}$)
* **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} 2x_1 \\ 2x_2 \end{pmatrix}$
* **Titik Awal:** $\mathbf{x}_0 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$. $\nabla f(\mathbf{x}_0) = \begin{pmatrix} 2 \\ 2 \end{pmatrix}$.
* **Inisialisasi QN:** $\mathbf{D}_0 = \mathbf{I}$ (Matriks Identitas).

### Penyelesaian (Langkah 1)

1.  **Langkah QN (Arah Pencarian):**
    $$\mathbf{p}_0 = -\mathbf{D}_0 \nabla f(\mathbf{x}_0) = - \mathbf{I} \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} -2 \\ -2 \end{pmatrix}$$
2.  **Pencarian Garis ($\alpha$):** Karena fungsi kuadratik, $\alpha_0=1$ adalah langkah optimal.
3.  **Posisi Baru:** $\mathbf{x}_1 = \mathbf{x}_0 + 1 \cdot \mathbf{p}_0 = \begin{pmatrix} 1 \\ 1 \end{pmatrix} + \begin{pmatrix} -2 \\ -2 \end{pmatrix} = \begin{pmatrix} -1 \\ -1 \end{pmatrix}$.
4.  **Gradien Baru:** $\nabla f(\mathbf{x}_1) = \begin{pmatrix} -2 \\ -2 \end{pmatrix}$.
5.  **Hitung Vektor Riwayat:**
    $$\mathbf{s}_0 = \mathbf{x}_1 - \mathbf{x}_0 = \begin{pmatrix} -2 \\ -2 \end{pmatrix}$$
    $$\mathbf{y}_0 = \nabla f(\mathbf{x}_1) - \nabla f(\mathbf{x}_0) = \begin{pmatrix} -2 \\ -2 \end{pmatrix} - \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} -4 \\ -4 \end{pmatrix}$$
6.  **Cek Kondisi Sekant:** Matriks Hessian asli adalah $\mathbf{H} = \begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix}$.
    $$\mathbf{H} \mathbf{s}_0 = \begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix} \begin{pmatrix} -2 \\ -2 \end{pmatrix} = \begin{pmatrix} -4 \\ -4 \end{pmatrix} = \mathbf{y}_0$$
    **Kesimpulan:** Kondisi sekant $\mathbf{B}_1 \mathbf{s}_0 = \mathbf{y}_0$ akan dipenuhi oleh $\mathbf{B}_1 = \mathbf{H}$. Dalam kasus kuadratik, BFGS akan menghasilkan $\mathbf{D}_k \approx \mathbf{H}^{-1}$ yang benar dalam $n$ langkah.

***

## 3. Contoh Menengah: BFGS 2D (Iterasi Matriks Penuh)

Ini adalah implementasi BFGS standar pada fungsi kuadratik yang terpelintir, di mana $\mathbf{D}_k$ tidak sama dengan $\mathbf{I}$ di awal.

### Masalah dan Rumus

* **Fungsi:** $f(\mathbf{x}) = 2x_1^2 + x_2^2 + x_1x_2$
* **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} 4x_1 + x_2 \\ x_1 + 2x_2 \end{pmatrix}$
* **Titik Awal:** $\mathbf{x}_0 = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$. $\nabla f(\mathbf{x}_0) = \begin{pmatrix} 9 \\ 4 \end{pmatrix}$.
* **Inisialisasi:** $\mathbf{D}_0 = \mathbf{I} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$.
* **Rumus BFGS (Invers Hessian $\mathbf{D}$):**
    $$\mathbf{D}_k = \left( \mathbf{I} - \frac{\mathbf{s}\mathbf{y}^T}{\mathbf{y}^T \mathbf{s}} \right) \mathbf{D}_{k-1} \left( \mathbf{I} - \frac{\mathbf{y}\mathbf{s}^T}{\mathbf{y}^T \mathbf{s}} \right) + \frac{\mathbf{s}\mathbf{s}^T}{\mathbf{y}^T \mathbf{s}}$$

### Penyelesaian (Iterasi 1)

1.  **Arah:** $\mathbf{p}_0 = -\mathbf{I} \nabla f(\mathbf{x}_0) = \begin{pmatrix} -9 \\ -4 \end{pmatrix}$.
2.  **Langkah:** Misalkan Line Search menemukan $\alpha_0 = 1/2$.
3.  **Posisi Baru:** $\mathbf{x}_1 = \begin{pmatrix} 2 \\ 1 \end{pmatrix} + 0.5 \begin{pmatrix} -9 \\ -4 \end{pmatrix} = \begin{pmatrix} -2.5 \\ -1.0 \end{pmatrix}$.
4.  **Gradien Baru:** $\nabla f(\mathbf{x}_1) = \begin{pmatrix} 4(-2.5) - 1 \\ -2.5 + 2(-1) \end{pmatrix} = \begin{pmatrix} -11 \\ -4.5 \end{pmatrix}$.
5.  **Hitung Riwayat ($\mathbf{s}_0, \mathbf{y}_0$):**
    $$\mathbf{s}_0 = \mathbf{x}_1 - \mathbf{x}_0 = \begin{pmatrix} -4.5 \\ -2.0 \end{pmatrix}$$
    $$\mathbf{y}_0 = \nabla f(\mathbf{x}_1) - \nabla f(\mathbf{x}_0) = \begin{pmatrix} -11 \\ -4.5 \end{pmatrix} - \begin{pmatrix} 9 \\ 4 \end{pmatrix} = \begin{pmatrix} -20 \\ -8.5 \end{pmatrix}$$
6.  **Hitung Komponen Skalar:** $\mathbf{y}_0^T \mathbf{s}_0 = (-20)(-4.5) + (-8.5)(-2.0) = 90 + 17 = 107$.
7.  **Pembaruan BFGS:** Substitusikan $\mathbf{s}_0, \mathbf{y}_0$, dan 107 ke dalam rumus BFGS untuk mendapatkan $\mathbf{D}_1$.

* **Kesimpulan:** Proses ini berlanjut. Karena fungsi kuadratik, $\mathbf{D}_k$ akan konvergen ke invers Hessian analitik $\mathbf{H}^{-1}$ dalam 2 langkah ($n$ langkah).

***

## 4. Contoh Sulit: BFGS Non-Kuadratik (Konsep Konvergensi Superlinear)

Dalam kasus non-kuadratik, BFGS sangat unggul karena menghasilkan konvergensi **superlinear** (lebih cepat dari linear, sedikit lebih lambat dari kuadratik murni).

### Masalah dan Rumus

* **Fungsi:** $f(\mathbf{x}) = \frac{1}{4}x_1^4 + \frac{1}{2}x_2^2 + x_1 - x_2$ (Cekung, non-kuadratik).
* **Gradien:** $\nabla f(\mathbf{x}) = \begin{pmatrix} x_1^3 + 1 \\ x_2 - 1 \end{pmatrix}$. Minimum di $\mathbf{x}^* = \begin{pmatrix} -1 \\ 1 \end{pmatrix}$.
* **Titik Awal:** $\mathbf{x}_0 = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$. $\nabla f(\mathbf{x}_0) = \begin{pmatrix} 2 \\ -1 \end{pmatrix}$.

### Analisis QN

1.  **Newton Murni:** Membutuhkan $\mathbf{H} = \begin{pmatrix} 3x_1^2 & 0 \\ 0 & 1 \end{pmatrix}$.
2.  **BFGS:** Menginisialisasi $\mathbf{D}_0 = \mathbf{I}$. Langkah pertama $\mathbf{p}_0$ adalah arah *steepest descent*.
3.  **Proses:** Karena fungsinya non-kuadratik, $\mathbf{D}_k$ akan terus diperbarui dan tidak pernah mencapai $\mathbf{H}^{-1}$ yang pasti (kecuali di $\mathbf{x}^*$). Namun, karena $\mathbf{D}_k$ menyerap informasi kelengkungan dari setiap Gradien, arah $\mathbf{p}_k$ menjadi semakin akurat.
4.  **Konvergensi:** Setelah beberapa langkah, $\mathbf{D}_k$ akan menjadi aproksimasi $\mathbf{H}^{-1}$ yang sangat baik, dan algoritma akan memasuki fase konvergensi superlinear, mirip dengan Metode Newton, tetapi tanpa biaya $O(n^3)$.

* **Kesimpulan:** BFGS adalah metode standar di sini karena mampu mengatasi perubahan kelengkungan (yang diwakili oleh $x_1^3$) hanya dengan informasi gradien.

***

## 5. Contoh Paling Sulit: L-BFGS (Optimasi Skala Besar)

Metode L-BFGS (Limited-memory BFGS) adalah solusi untuk masalah di mana $n$ (jumlah variabel) terlalu besar (misalnya, $n > 10,000$).

### Masalah dan Rumus

* **Masalah:** Minimalkan fungsi kerugian di *Deep Learning* (misalnya, fungsi kerugian logistik dengan $n=10^5$ *feature*).
* **Keterbatasan Memori:** Matriks $\mathbf{D}$ berukuran $10^5 \times 10^5$ tidak dapat disimpan di memori (sekitar 80GB).

### Komponen L-BFGS

1.  **Penyimpanan Riwayat Terbatas:** Matriks $\mathbf{D}_k$ tidak disimpan. Hanya **$m$ pasang** vektor riwayat ($\mathbf{s}_i, \mathbf{y}_i$) terakhir yang disimpan ($m \approx 5$ hingga 20).
    $$R_k = \{ (\mathbf{s}_i, \mathbf{y}_i) \mid i = k-m, \dots, k-1 \}$$
2.  **Algoritma Dua *Loop* (Two-Loop Recursion):** Arah pencarian $\mathbf{p}_k = -\mathbf{D}_k \nabla f(\mathbf{x}_k)$ dihitung melalui prosedur rekursif yang hanya melibatkan perkalian matriks-vektor dengan vektor-vektor riwayat yang tersimpan.
    $$\mathbf{p}_k = \text{Two-Loop-Recursion}(\nabla f(\mathbf{x}_k), R_k)$$

### Keunggulan

* **Biaya Memori:** Biaya memori dikurangi dari $O(n^2)$ menjadi $O(mn)$.
* **Biaya Komputasi:** Biaya untuk menghitung $\mathbf{p}_k$ berkurang dari $O(n^3)$ menjadi $O(mn)$.
* **Konvergensi:** Tetap mempertahankan konvergensi superlinear yang sangat cepat, menjadikannya standar emas untuk optimasi dengan $n$ yang besar, seperti di *machine learning*.

# Ilustrasi Quasi Newton

Untuk mengilustrasikan **Metode Kuasi-Newton (BFGS)** dalam 3D, kita perlu menunjukkan bagaimana jalur optimasi (path) di permukaan fungsi berbeda dari Gradient Descent (GD) dan Metode Newton murni.

Berikut adalah 5 contoh visualisasi 3D yang menunjukkan aspek berbeda dari Metode Kuasi-Newton, diurutkan berdasarkan kompleksitas fungsi yang divisualisasikan.

### Catatan Penting

Karena BFGS adalah metode iteratif yang kompleks, semua contoh di bawah menggunakan fungsi simulasi `bfgs_optimizer` yang sudah ditentukan, yang akan melacak jalur $(\mathbf{x}, f(\mathbf{x}))$ untuk keperluan visualisasi.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize # Menggunakan implementasi BFGS SciPy yang stabil

# --- FUNGSI OPTIMASI UMUM (Menggunakan SciPy untuk stabilitas BFGS) ---



````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize # Menggunakan implementasi BFGS SciPy yang stabil

# --- FUNGSI OPTIMASI UMUM (Menggunakan SciPy untuk stabilitas BFGS) ---
def bfgs_optimizer_tracker(f, x_start, grad_func, max_iter=15):
    path = [np.array(x_start)]

    def callback(xk):
        path.append(xk.copy())

    result = minimize(f, x_start, method='BFGS', jac=grad_func,
                      callback=callback, options={'maxiter': max_iter})

    # Pastikan path akhir ditambahkan jika belum
    if not np.array_equal(path[-1], result.x):
        path.append(result.x)

    return np.array(path)

# --- FUNGSI PLOTTING 3D UMUM ---
def plot_3d_path(f, path, X1, X2, Z, title, cmap='viridis', elev=30, azim=120):
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Plot Permukaan
    ax.plot_surface(X1, X2, Z, cmap=cmap, alpha=0.7, rstride=5, cstride=5)

    # Plot Jalur BFGS
    path_z = np.array([f(p) for p in path])
    ax.plot(path[:, 0], path[:, 1], path_z, color='red', marker='o', markersize=6,
            label='Jalur Iterasi BFGS', linewidth=3)

    # Titik Awal & Akhir
    ax.scatter(path[0, 0], path[0, 1], path_z[0], color='black', s=100, label='Titik Awal')
    ax.scatter(path[-1, 0], path[-1, 1], path_z[-1], color='gold', marker='*', s=400, label='Titik Optimum')

    ax.set_xlabel('$x_1$'); ax.set_ylabel('$x_2$'); ax.set_zlabel('$f(x)$')
    ax.set_title(title, fontsize=14)
    ax.legend()
    ax.view_init(elev=elev, azim=azim)
    plt.show()

# --- Fungsi yang Digunakan ---
# (Rosenbrock akan didefinisikan di C5)
````

````python
def f_parabola_miring(x): return 2*x[0]**2 + x[1]**2 + x[0]*x[1]
def grad_parabola_miring(x): return np.array([4*x[0] + x[1], x[0] + 2*x[1]])

x1_range, x2_range = np.linspace(-1, 4, 50), np.linspace(-3, 3, 50)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f_parabola_miring([X1, X2])

path = bfgs_optimizer_tracker(f_parabola_miring, [3.0, 1.0], grad_parabola_miring, max_iter=2)

plot_3d_path(f_parabola_miring, path, X1, X2, Z,
             "C1: BFGS pada Parabola Miring (Konvergensi 2 Langkah)",
             cmap='plasma')
````

````python
def f_kuadratik_power(x): return x[0]**4 + x[1]**4 + x[0]**2 + x[1]**2
def grad_kuadratik_power(x): return np.array([4*x[0]**3 + 2*x[0], 4*x[1]**3 + 2*x[1]])

x1_range, x2_range = np.linspace(-2.5, 2.5, 50), np.linspace(-2.5, 2.5, 50)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f_kuadratik_power([X1, X2])
Z[Z > 100] = 100 # Pembatasan untuk visual

path = bfgs_optimizer_tracker(f_kuadratik_power, [2.0, 2.0], grad_kuadratik_power, max_iter=5)

plot_3d_path(f_kuadratik_power, path, X1, X2, Z,
             "C2: BFGS pada Permukaan Non-Kuadratik Mulus",
             cmap='cividis')
````

````python
def f_rosen(x): return 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2
def grad_rosen(x): return np.array([-400*x[0]*(x[1] - x[0]**2) - 2*(1 - x[0]), 200*(x[1] - x[0]**2)])

x1_range, x2_range = np.linspace(-1.5, 1.5, 100), np.linspace(-0.5, 2.5, 100)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f_rosen([X1, X2])
Z[Z > 100] = 100

path = bfgs_optimizer_tracker(f_rosen, [-1.2, 1.0], grad_rosen, max_iter=15)

plot_3d_path(f_rosen, path, X1, X2, Z,
             "C3: BFGS pada Lembah Rosenbrock (Navigasi Kelengkungan)",
             cmap='terrain', elev=20, azim=-140)
````

````python
# Fungsi GD untuk perbandingan
def gd_optimizer_tracker(f, x_start, grad_func, alpha=0.001, max_iter=20):
    x = np.array(x_start, dtype=float)
    path = [x.copy()]
    for _ in range(max_iter):
        g = grad_func(x)
        x = x - alpha * g # Langkah kecil konstan
        path.append(x.copy())
    return np.array(path)

# Gunakan Rosenbrock karena perbedaannya paling jelas
path_bfgs = bfgs_optimizer_tracker(f_rosen, [0.0, 0.0], grad_rosen, max_iter=10)
path_gd = gd_optimizer_tracker(f_rosen, [0.0, 0.0], grad_rosen, alpha=0.002, max_iter=300) # GD butuh ratusan langkah

# Plot Komparatif
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='coolwarm', alpha=0.6)

# Plot Jalur BFGS (Hanya 10 Langkah)
path_bfgs_z = np.array([f_rosen(p) for p in path_bfgs])
ax.plot(path_bfgs[:, 0], path_bfgs[:, 1], path_bfgs_z, color='red', marker='o', markersize=5, linewidth=3, label='BFGS (Cepat)')

# Plot Jalur Gradient Descent (300 Langkah, Lambat)
path_gd_z = np.array([f_rosen(p) for p in path_gd[:50]]) # Batasi langkah GD agar plot tidak penuh
ax.plot(path_gd[:50, 0], path_gd[:50, 1], path_gd_z, color='blue', marker='', linestyle='--', linewidth=1.5, label='GD (Zig-Zag)')

ax.set_title("C4: Komparasi BFGS (Merah) vs. Gradient Descent (Biru)")
ax.view_init(elev=20, azim=120)
ax.legend()
plt.show()
````

````python
def f_six_hump(x):
    x1, x2 = x
    return (4 - 2.1*x1**2 + x1**4/3) * x1**2 + x1*x2 + (-4 + 4*x2**2) * x2**2

# Gradiennya sangat kompleks, kita gunakan optimasi numerik SciPy tanpa jac explisit
# SciPy akan menggunakan aproksimasi gradien (finite differences)

x1_range, x2_range = np.linspace(-2, 2, 50), np.linspace(-1, 1, 50)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f_six_hump([X1, X2])

# Menggunakan BFGS tanpa Jacobian eksplisit (SciPy akan menggunakan aproksimasi FD)
# Ini adalah praktek umum ketika turunan sulit
path = bfgs_optimizer_tracker(f_six_hump, [1.0, 0.0], grad_func=None, max_iter=10)

plot_3d_path(f_six_hump, path, X1, X2, Z,
             "C5: BFGS pada Six-Hump Camel (Optimum Lokal Ganda)",
             cmap='coolwarm', elev=45, azim=15)
````

## Kata Kunci Utama Kuasi-Newton

### 1. Aproksimasi Hessian ($\mathbf{D}_k$ atau $\mathbf{B}_k$)
* **Inti:** Kuasi-Newton tidak pernah menghitung Matriks **Hessian** asli ($\mathbf{H}$, turunan kedua) yang mahal ($O(n^2)$ elemen dan $O(n^3)$ penyelesaian).
* **Pengganti:** Metode ini menggunakan Matriks Aproksimasi ($\mathbf{B}_k$) atau Matriks Aproksimasi Invers Hessian ($\mathbf{D}_k$) yang diperbarui di setiap iterasi.
* **Mudah Dipahami:** Ini seperti menggunakan "peta lama yang diperbarui secara *real-time*" daripada membuat peta baru yang mahal dari awal di setiap langkah.

***

### 2. Kondisi Sekant (Secant Condition)
* **Inti:** Ini adalah syarat matematis yang harus dipenuhi oleh matriks aproksimasi ($\mathbf{B}_k$) agar valid. Kondisi ini menuntut agar $\mathbf{B}_k$ bekerja seperti Hessian asli **sepanjang langkah terakhir ($\mathbf{s}_{k-1}$)**.
    $$\mathbf{B}_k \mathbf{s}_{k-1} = \mathbf{y}_{k-1}$$
    di mana $\mathbf{s}_{k-1}$ adalah perubahan posisi dan $\mathbf{y}_{k-1}$ adalah perubahan Gradien.
* **Mudah Dipahami:** Aturan ini memastikan bahwa matriks aproksimasi (**$\mathbf{B}_k$ telah "belajar" kelengkungan rata-rata** dari fungsi pada pergerakan terakhir, sehingga langkah berikutnya bisa lebih cerdas.

***

### 3. Pembaruan Rank-Dua (Rank-Two Update)
* **Inti:** Rumus yang digunakan untuk mengubah $\mathbf{D}_{k-1}$ menjadi $\mathbf{D}_k$ adalah pembaruan *rank-dua*. Ini berarti matriks baru dihasilkan dengan menambahkan dua matriks rank satu.
* **Keuntungan:** Pembaruan rank-dua adalah metode paling efisien (hanya $O(n^2)$ operasi) untuk memperbarui $\mathbf{D}_k$ sambil **menjaga sifat positif definit** (memastikan $\mathbf{D}_k$ selalu mengarah ke minimum, bukan maksimum atau pelana).
* **Mudah Dipahami:** Daripada menghitung ulang seluruh matriks $n \times n$ (yang sangat besar), kita hanya melakukan *koreksi kecil* (dua perbaikan) pada matriks lama berdasarkan informasi Gradien terbaru.

***

### 4. BFGS (Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno)
* **Inti:** Ini adalah **algoritma Kuasi-Newton paling populer** dan paling andal. BFGS adalah pembaruan rank-dua yang dirancang untuk secara langsung memperbarui $\mathbf{D}_k$ (aproksimasi Invers Hessian).
* **Keunggulan:** BFGS secara inheren menjaga $\mathbf{D}_k$ tetap positif definit, yang sangat penting untuk stabilitas optimasi.
* **Mudah Dipahami:** BFGS adalah "nama merek" yang dominan. Ketika orang mengatakan Kuasi-Newton, mereka hampir selalu merujuk pada BFGS.

***

### 5. Konvergensi Superlinear (Superlinear Convergence)
* **Inti:** Kecepatan konvergensi QN berada di antara konvergensi *linear* (lambat, seperti Gradient Descent) dan *kuadratik* (sangat cepat, seperti Newton murni).
* **Definisi:** Error berkurang lebih cepat daripada faktor linear manapun.
* **Mudah Dipahami:** QN "hampir secepat" Metode Newton. Ini memberi kita kecepatan tinggi tanpa biaya $O(n^3)$ yang mahal per iterasi. Kecepatan ini tercapai karena $\mathbf{D}_k$ menjadi aproksimasi yang sangat baik untuk $\mathbf{H}^{-1}$ saat mendekati solusi.

***

### 6. L-BFGS (Limited-memory BFGS)
* **Inti:** Varian BFGS yang dirancang untuk optimasi **skala besar** (ribuan hingga jutaan variabel, $n$ sangat besar).
* **Modifikasi:** L-BFGS **tidak menyimpan Matriks Aproksimasi $\mathbf{D}_k$** secara penuh di memori. Sebaliknya, ia hanya menyimpan sejumlah kecil (*m* pasang) vektor riwayat langkah ($\mathbf{s}_i, \mathbf{y}_i$) terakhir.
* **Mudah Dipahami:** Ini adalah versi "hemat memori." Matriks utuh terlalu besar (misalnya, 80 GB), jadi L-BFGS menggunakan metode rekursif yang hanya membutuhkan memori $O(mn)$ dan tetap mempertahankan kecepatan tinggi.

Tentu! Untuk materi yang detail dan lengkap, berikut adalah lebih banyak kata kunci dari Metode Kuasi-Newton (QN), dibagi berdasarkan peran matematis dan proseduralnya.

***


### I. Komponen Matematis dan Prosedural

#### 7. Vektor Perbedaan ($\mathbf{s}$ dan $\mathbf{y}$)
* **Inti:** Vektor $\mathbf{s}_{k-1}$ (perbedaan posisi) dan $\mathbf{y}_{k-1}$ (perbedaan gradien) adalah **input data** yang digunakan untuk memperbarui matriks aproksimasi $\mathbf{D}_k$.
* **$\mathbf{s}_{k-1}$ (Langkah Posisi):** $\mathbf{x}_k - \mathbf{x}_{k-1}$.
* **$\mathbf{y}_{k-1}$ (Langkah Gradien):** $\nabla f(\mathbf{x}_k) - \nabla f(\mathbf{x}_{k-1})$.
* **Mudah Dipahami:** Ini adalah **"memori"** dari dua langkah terakhir yang digunakan QN untuk menghitung bentuk kelengkungan.

#### 8. Persamaan Kompatibilitas (Compatibility Equation)
* **Inti:** Nama lain yang lebih formal untuk **Kondisi Sekant** ($\mathbf{B}_k \mathbf{s}_{k-1} = \mathbf{y}_{k-1}$). Istilah ini menekankan bahwa matriks $\mathbf{B}_k$ harus **kompatibel** dengan data turunan yang diamati.
* **Mudah Dipahami:** Ini adalah "persamaan wajib" yang menjamin matriks aproksimasi *secara matematis masuk akal* berdasarkan pergerakan Gradien yang terjadi.

#### 9. Sifat Positif Definit (Positive Definiteness)
* **Inti:** Syarat bahwa matriks aproksimasi $\mathbf{D}_k$ harus **positif definit**. Jika $\mathbf{D}_k$ positif definit, maka arah pencarian $\mathbf{p}_k = -\mathbf{D}_k \nabla f(\mathbf{x}_k)$ **dijamin sebagai arah penurunan (descent direction)**.
* **Mudah Dipahami:** Ini menjamin bahwa setiap langkah yang diambil oleh QN pasti menuju ke "bawah" menuju minimum, dan tidak akan melompat ke maksimum atau ke samping. BFGS dirancang secara khusus untuk mempertahankan sifat ini.

#### 10. *Line Search* (Pencarian Garis)
* **Inti:** Prosedur internal yang digunakan untuk menemukan **panjang langkah optimal ($\alpha_k$)** di sepanjang arah pencarian $\mathbf{p}_k$.
* **Tujuan:** Memastikan penurunan nilai fungsi yang cukup (kondisi Armijo) dan menjaga agar $\alpha_k$ tidak terlalu kecil (kondisi Wolfe).
* **Mudah Dipahami:** Walaupun arahnya sudah bagus (diberikan oleh $\mathbf{D}_k$), kita tetap perlu Line Search untuk **mengendalikan seberapa jauh** kita melangkah, mencegah overshooting dan menjaga stabilitas.

***

### II. Perbandingan dan Konteks

#### 11. Metode Orde Kedua (Second-Order Method)
* **Inti:** QN diklasifikasikan sebagai metode orde kedua, sama seperti Metode Newton.
* **Alasan:** QN menggunakan informasi kelengkungan fungsi (yang merupakan turunan orde kedua) melalui aproksimasi $\mathbf{D}_k$.
* **Mudah Dipahami:** Ini adalah mengapa QN sangat cepatâ€”ia menggunakan informasi bentuk (kelengkungan), tidak hanya informasi kemiringan (lereng) seperti metode orde pertama.

#### 12. Metode Kuasi-Newton Skala Besar (Large-Scale QN)
* **Inti:** Istilah umum yang mencakup **L-BFGS** dan varian lainnya. Digunakan ketika jumlah variabel $n$ sangat besar ($n > 1000$).
* **Ciri Khas:** Pengurangan dramatis kebutuhan memori dari $O(n^2)$ menjadi $O(mn)$.
* **Mudah Dipahami:** Istilah payung untuk semua teknik QN yang dioptimalkan untuk *big data* dan *machine learning*.

#### 13. DFP (Davidonâ€“Fletcherâ€“Powell)
* **Inti:** Metode Kuasi-Newton bersejarah dan penting secara teoritis. DFP adalah pembaruan rank-dua yang merupakan **pasangan komplementer** dari BFGS.
* **Hubungan:** DFP memperbarui $\mathbf{D}_k$ secara langsung, sedangkan BFGS memperbarui $\mathbf{B}_k$ (aproksimasi Hessian) yang kemudian menghasilkan pembaruan $\mathbf{D}_k$. Kedua rumus pembaruan terkait melalui rumus matriks Sherman-Morrison.
* **Mudah Dipahami:** Ini adalah "pendahulu yang penting" bagi BFGS, menunjukkan bahwa ada lebih dari satu cara untuk memperbarui matriks aproksimasi.



## Kata Kunci Metode Newton (Optimasi)

### I. Konsep Dasar dan Tujuan

| Kata Kunci | Penjelasan Detail & Mudah Dipahami |
| :--- | :--- |
| **Minimum Lokal** | Titik di mana nilai fungsi $f(\mathbf{x})$ paling rendah **dibandingkan dengan titik-titik di sekitarnya**. Metode Newton bertujuan menemukan titik ini ($\mathbf{x}^*$), di mana kemiringan menjadi nol. |
| **Titik Kritis** | Titik $\mathbf{x}$ di mana **Gradien sama dengan nol** ($\nabla f(\mathbf{x}) = \mathbf{0}$). Metode Newton dirancang untuk menemukan titik kritis ini, yang bisa berupa minimum, maksimum, atau titik pelana. |
| **Metode Orde Kedua** | Metode optimasi yang menggunakan turunan **orde kedua** (Matriks Hessian) selain turunan orde pertama (Gradien). Ini menjadikannya metode yang *cerdas* karena mempertimbangkan bentuk atau kelengkungan fungsi. |

***

### II. Komponen Matematis Inti

| Kata Kunci | Penjelasan Detail & Mudah Dipahami |
| :--- | :--- |
| **Gradien ($\nabla f$)** | Vektor turunan parsial pertama. Menunjukkan **arah paling curam** (lereng) dari fungsi. Digunakan untuk mengetahui ke mana harus bergerak. |
| **Matriks Hessian ($\mathbf{H}$)** | Matriks turunan parsial kedua. Menunjukkan **kelengkungan** (konveksitas/cekungan) fungsi. Ini yang membedakan minimum, maksimum, dan titik pelana. |
| **Aproksimasi Kuadratik** | Ide dasar Newton: fungsi yang rumit diaproksimasi di sekitar titik $\mathbf{x}_k$ dengan **fungsi kuadratik (parabola)** yang mudah diminimalkan. Meminimalkan parabola ini memberikan langkah Newton. |
| **Langkah Newton ($\mathbf{p}_k$)** | Vektor arah yang ditemukan dengan menyelesaikan sistem $\mathbf{H}_k \mathbf{p}_k = -\nabla f_k$. Arah ini membawa kita **langsung ke dasar parabola** aproksimasi dalam satu langkah. |

***

### III. Sifat dan Kinerja

| Kata Kunci | Penjelasan Detail & Mudah Dipahami |
| :--- | :--- |
| **Konvergensi Kuadratik** | Kecepatan konvergensi yang sangat cepat. Pada setiap iterasi yang berhasil, **jumlah digit akurat pada solusi akan berlipat ganda**. Terjadi hanya ketika iterasi sudah dekat dengan solusi. |
| **Konvergensi Lokal** | Jaminan keberhasilan Newton **hanya berlaku jika** titik awal ($\mathbf{x}_0$) berada cukup dekat dengan minimum yang dicari ($\mathbf{x}^*$). Jika terlalu jauh, metode bisa gagal (divergen). |
| **Biaya Komputasi $O(n^3)$** | Keterbatasan utama: biaya yang sangat tinggi per langkah. Biaya didominasi oleh perhitungan dan penyelesaian sistem yang melibatkan Hessian $n \times n$, yang memakan waktu sebanding dengan $n^3$ (pangkat tiga) dari jumlah variabel. |
| **Titik Pelana (Saddle Point)** | Titik kritis di mana fungsi cekung pada satu arah dan cembung pada arah lain (Hessian indefinit). Metode Newton dapat konvergen ke titik ini jika titik awal diarahkan ke sana. |

***

### IV. Modifikasi dan Stabilitas

| Kata Kunci | Penjelasan Detail & Mudah Dipahami |
| :--- | :--- |
| **Metode Newton Terhambat (Damped Newton)** | Modifikasi yang paling umum untuk mengatasi masalah konvergensi. Langkah Newton $\mathbf{p}_k$ dikalikan dengan faktor panjang langkah $\alpha_k \in (0, 1]$. |
| ***Line Search*** | Prosedur yang digunakan untuk menemukan faktor $\alpha_k$ yang tepat. Tujuannya adalah **memastikan** bahwa nilai fungsi **pasti menurun** pada setiap langkah, menjamin stabilitas global. |
| **Konvergensi Global** | Tujuan modifikasi: agar metode dapat dijamin konvergen ke minimum (atau titik kritis) **terlepas dari titik awal** ($\mathbf{x}_0$). Dicapai dengan menggunakan Line Search. |
| **Hessian Modifikasi** | Teknik yang digunakan ketika Hessian $\mathbf{H}_k$ tidak positif definit (misalnya, di titik pelana). Matriks diubah sedikit (dibuat lebih positif) agar arah $\mathbf{p}_k$ tetap menjadi **arah penurunan** yang valid. |

## V. Interpretasi Geometris (Mengapa Newton Cerdas?) ðŸ§ 

Metode Newton tidak hanya cepat; ia bekerja karena memiliki pemahaman geometris yang superior terhadap bentuk fungsi dibandingkan Gradient Descent.

### A. Peran Geometris Gradien

* **Gradien ($\nabla f$):** Menunjukkan **arah paling curam (Steepest Descent)**. Jika kita hanya menggunakan gradien, langkah kita selalu tegak lurus (ortogonal) terhadap garis kontur (seperti mengambil jalur tercuram saat menuruni bukit, yang seringkali tidak efisien).

### B. Peran Geometris Hessian

* **Hessian ($\mathbf{H}$):** Memberikan informasi tentang **Kelengkungan (Curvature)**. Hessian mendefinisikan bentuk lokal fungsi.
    * Pada fungsi **bulat sempurna** (kontur melingkar), Hessian adalah matriks skalar.
    * Pada fungsi **memanjang** (kontur elips/lembah sempit), Hessian menunjukkan bagaimana fungsi terpelintir dan seberapa curamnya di sepanjang sumbu yang berbeda.

### C. Aksi Langkah Newton

* **Langkah Newton ($\mathbf{p}_k = -\mathbf{H}^{-1} \nabla f$):** $\mathbf{H}^{-1}$ bertindak sebagai **pengubah skala (scaling factor)** dan **rotator**. Ia mengambil arah Steepest Descent yang "naif" ($\nabla f$) dan **mengoreksinya** berdasarkan kelengkungan yang diukur oleh $\mathbf{H}$.
    * **Hasil:** Langkah Newton mengarah langsung ke minimum lokal *parabola* (aproksimasi kuadratik) dalam satu langkah. Secara visual, langkah Newton tidak selalu tegak lurus terhadap kontur, tetapi ia memotongnya dengan sudut yang dihitung untuk efisiensi maksimum.

## VI. Kondisi Orde Kedua (Validasi Minimum) âœ…

Dalam optimasi, menemukan titik kritis saja tidak cukup; kita harus memverifikasi bahwa titik tersebut adalah minimum.

### A. Kondisi Orde Pertama yang Diperlukan (FOCN)
* **Kondisi:** $\nabla f(\mathbf{x}^*) = \mathbf{0}$.
* **Fungsi:** Ini adalah syarat **wajib** bagi $\mathbf{x}^*$ untuk menjadi titik kritis (bisa minimum, maksimum, atau pelana).

### B. Kondisi Orde Kedua yang Cukup (SOSC)
* **Kondisi:** $\nabla f(\mathbf{x}^*) = \mathbf{0}$ **DAN** Matriks Hessian $\mathbf{H}(\mathbf{x}^*)$ harus **positif definit**.
* **Makna Positif Definit:** Semua *eigenvalue* dari $\mathbf{H}(\mathbf{x}^*)$ harus positif. Ini secara geometris berarti fungsi **cekung ke atas** di semua arah di sekitar $\mathbf{x}^*$.
* **Fungsi:** Ini adalah syarat yang **cukup** untuk menjamin bahwa titik kritis $\mathbf{x}^*$ benar-benar adalah **minimum lokal** (bukan maksimum atau pelana).

### C. Keterbatasan Newton Terhadap SOSC

* **Keterbatasan:** Metode Newton murni hanya bertujuan mencapai FOCN ($\nabla f = 0$). Metode ini tidak secara inheren menjamin atau memeriksa SOSC.
* **Implikasi:** Jika Newton konvergen ke titik di mana Hessian *tidak* positif definit (indefinit atau negatif definit), modifikasi seperti **Hessian Modifikasi** atau **Trust Region** diperlukan untuk mengubah arah pencarian.
